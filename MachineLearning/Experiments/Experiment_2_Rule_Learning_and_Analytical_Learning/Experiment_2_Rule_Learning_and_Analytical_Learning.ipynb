## üìñ **Background & Theory**

### 1Ô∏è‚É£ CN2 ‚Äì Example-Based Learning

The **CN2 algorithm** learns a list of **IF‚ÄìTHEN rules** that classify data points.

#### üîç Working Steps:

1. Start with all labeled examples.
2. Generate **candidate conditions** (combinations of attributes).
3. Evaluate each rule using **entropy** or **Laplace accuracy**.
4. Keep rules that cover *mostly positive examples*.
5. Repeat until all examples are classified by some rule.

**Output Example:**

```

IF fever = VeryHigh AND breathing = Severe THEN diagnosis = COVID-19
IF fever = Mild AND cough = Mild THEN diagnosis = Common\_Cold

```

### 2Ô∏è‚É£ FOIL ‚Äì First-Order Inductive Learning

The **FOIL algorithm** learns **logical rules** that involve relationships between entities.

#### üîç Working Steps:

1. Begin with positive and negative examples of a target concept.
2. Generate candidate rules using known predicates (e.g., Parent, Brother).
3. Evaluate the rule‚Äôs **information gain** to decide which literal to add.
4. Continue refining until all positive examples are explained.

**Output Example:**

```

Grandparent(X, Y) :- Parent(X, Z) AND Parent(Z, Y)
Uncle(X, Y) :- Brother(X, Z) AND Parent(Z, Y)

```

---

## üìò **Mathematical Foundations (Simplified)**

### üß© CN2 Rule Evaluation ‚Äì *Entropy*

**Entropy** measures how mixed or pure a set of examples is.

* **Entropy = 0** ‚Üí All examples belong to one class (perfectly pure).
* **High Entropy** ‚Üí Examples are mixed between classes (uncertain).

**Conceptually:**

```

Entropy(S) = -p(+) \* log2(p(+)) - p(-) \* log2(p(-))

```

Where:

* `p(+)` = proportion of positive examples
* `p(-)` = proportion of negative examples

üß† **Intuition:**
The CN2 algorithm prefers rules that **lower entropy**, i.e., that clearly separate one class from the other.

### üß© FOIL Rule Evaluation ‚Äì *Information Gain*

**Information Gain** measures how much a rule improves when adding a new condition (literal).

**Conceptually:**

```

Gain = p1 \* [ log2(p1 / (p1 + n1)) - log2(p0 / (p0 + n0)) ]

```

Where:

* `(p0, n0)` = positives and negatives before adding the new literal
* `(p1, n1)` = positives and negatives after adding the literal

üß† **Intuition:**
The higher the gain, the more the new condition improves the rule ‚Äî helping FOIL build **smarter, more accurate rules**.

A **choke cut** is a controlled reduction of flow from a well by adjusting the choke valve. This process is used to manage wellbore pressure, regulate production rates, and prevent formation damage. By gradually restricting the flow, operators can safely bring a well under control, minimize pressure surges, and reduce the risk of equipment failure or blowouts.

Choke cuts are typically performed during well start-up, testing, or when changing production conditions. The procedure involves incrementally closing the choke valve to achieve the desired flow rate and pressure, while closely monitoring well parameters to ensure safe and stable operations.

If a choke cut is not performed correctly, several safety hazards may arise:

- **Well Blowouts:** Improper choke management can lead to uncontrolled flow, increasing the risk of blowouts, which are dangerous and can cause loss of life, equipment, and environmental damage.
- **Equipment Failure:** Excessive flow rates may exceed equipment design limits, leading to valve or pipeline ruptures.
- **Formation Damage:** Incorrect choke settings can cause rapid pressure drops, resulting in sand production or water coning, which may damage the wellbore and surface facilities.
- **Fire and Explosion:** Uncontrolled hydrocarbon release increases the risk of fire or explosion, especially in the presence of ignition sources.
- **Environmental Pollution:** Spills or leaks due to poor choke control can contaminate soil and water, leading to regulatory penalties and cleanup costs.

Proper choke management is essential to maintain safe operations and protect personnel, equipment, and the environment.

## üîç Section 2: The Sequential Covering Algorithm (CN2)

Sequential Covering is the general strategy used by algorithms like CN2 (Classification with New Generation) to learn these rules. It works by focusing on one class (or one part of the data) at a time until all examples are 'covered' by a rule.

The core idea is: **"Learn a good rule, remove the examples it covers, and repeat."**

1.  **Select a Target Class:** Pick one class to learn rules for (e.g., 'COVID-19').
2.  **Learn the Best Rule:** Start with a very general rule (`IF TRUE THEN COVID-19`) and iteratively **add conditions** to the IF part (e.g., add `Fever = VeryHigh`) until the rule becomes highly **accurate** (predicts mostly positive examples for that class) and still covers a good number of examples (**coverage**).
    * **CN2** uses a statistical measure (like entropy or Laplace accuracy) to find the best conditions.

3.  **Cover and Eliminate:** All training examples correctly classified by this new rule are removed from the training set.
4.  **Repeat:** Go back to Step 2 and find the next best rule for the remaining, uncovered examples.
5.  **Stop:** The process stops when no more rules can be found with acceptable accuracy, or all examples of the target class have been covered.
6.  **Default Rule:** A final **default rule** is added to classify any remaining, uncovered examples (often to the majority class of the remaining examples, or simply the most frequent class overall).

### Why this is Example-Based

This method is "example-based" because the search for rules is driven entirely by the **positive and negative examples** in the training data. The algorithm literally *looks at* the remaining uncovered examples to guide the search for the next best rule, ensuring every training example contributes to the final rule set.

````python
import json
import pandas as pd
import matplotlib.pyplot as plt
import copy
from typing import List, Dict, Any, Tuple

# ============================================================================
# 1. HELPER FUNCTIONS FOR RULE EVALUATION
# ============================================================================

def covers_example(rule_antecedent: Dict[str, str], example: Dict[str, Any]) -> bool:
    """Checks if a single training example is covered by a rule's antecedent."""
    for attr, val in rule_antecedent.items():
        # Check for matching attribute value
        if example.get(attr) != val:
            return False
    return True

def get_covered_examples(rule_antecedent: Dict[str, str], examples: List[Dict[str, Any]]) -> List[Dict[str, Any]]:
    """Returns the subset of examples covered by the rule's antecedent."""
    covered = []
    for example in examples:
        if covers_example(rule_antecedent, example):
            covered.append(example)
    return covered

def calculate_rule_accuracy(rule_antecedent: Dict[str, str], consequent_class: str,
                            examples: List[Dict[str, Any]], target_name: str) -> float:
    """
    Calculates Laplace corrected accuracy: (Positives Covered + 1) / (Total Covered + Num Classes).
    """
    covered_examples = get_covered_examples(rule_antecedent, examples)
    total_covered = len(covered_examples)

    all_classes = set(ex.get(target_name) for ex in examples)
    num_classes = len(all_classes)

    positives = sum(1 for ex in covered_examples if ex.get(target_name) == consequent_class)

    if total_covered == 0:
        return 0.0

    laplace_acc = (positives + 1) / (total_covered + num_classes)
    return laplace_acc

def format_rule(rule_antecedent: Dict[str, str], consequent_class: str, target_name: str) -> str:
    """Formats the rule for pretty printing: IF <conditions> THEN <class>."""
    if not rule_antecedent:
        antecedent_str = 'TRUE'
    else:
        conditions = [f"({attr} = {val})" for attr, val in rule_antecedent.items()]
        antecedent_str = ' AND '.join(conditions)

    return f"IF {antecedent_str} THEN {target_name} = {consequent_class}"


print('‚úÖ Helper functions for Sequential Covering defined successfully!')
print(' ¬†‚Ä¢ covers_example() - Checks if an example matches a rule.')
print(' ¬†‚Ä¢ get_covered_examples() - Returns all examples a rule covers.')
print(' ¬†‚Ä¢ calculate_rule_accuracy() - Measures rule quality (Accuracy).')
print(' ¬†‚Ä¢ format_rule() - Pretty prints the rule.')
````

````python
# ============================================================================
#  SEQUENTIAL COVERING ALGORITHM IMPLEMENTATION
# ============================================================================

def find_best_rule(current_examples: List[Dict[str, Any]], target_class: str,
                   attribute_names: List[str], target_name: str) -> Tuple[Dict[str, str], float]:
    """
    Greedy search for the best rule. Implements a tie-breaker: prefer the most specific
    rule (most conditions) when accuracy is equal, allowing for longer rules.
    """

    current_antecedent = {}
    best_antecedent = {}
    min_coverage = 1

    initial_accuracy = calculate_rule_accuracy({}, target_class, current_examples, target_name)
    best_accuracy = initial_accuracy

    while True:
        best_specialized_antecedent = None
        best_specialized_accuracy = -1.0

        new_candidate_rules = []
        for attr in attribute_names:
            if attr not in current_antecedent:
                unique_values = set(ex.get(attr) for ex in current_examples)
                for val in unique_values:
                    if val is not None:
                        new_antecedent = current_antecedent.copy()
                        new_antecedent[attr] = val
                        new_candidate_rules.append(new_antecedent)

        # B. Evaluate all new candidate rules and find the best single specialization
        for new_antecedent in new_candidate_rules:
            current_coverage = len(get_covered_examples(new_antecedent, current_examples))

            if current_coverage >= min_coverage:
                new_accuracy = calculate_rule_accuracy(
                    new_antecedent, target_class, current_examples, target_name
                )

                # Select the best specialization based on accuracy and specificity (tie-breaker)
                if new_accuracy > best_specialized_accuracy:
                    best_specialized_accuracy = new_accuracy
                    best_specialized_antecedent = new_antecedent
                elif new_accuracy == best_specialized_accuracy and len(new_antecedent) > len(best_specialized_antecedent or {}):
                    # Tie-breaker: prefer the more specific rule
                    best_specialized_accuracy = new_accuracy
                    best_specialized_antecedent = new_antecedent

        # C. Decide whether to keep the specialization
        current_best_acc_for_comparison = best_accuracy

        # 1. Check for strict improvement
        if best_specialized_antecedent and best_specialized_accuracy > current_best_acc_for_comparison:
            # Found a strictly better rule
            current_antecedent = best_specialized_antecedent
            best_accuracy = best_specialized_accuracy
            best_antecedent = current_antecedent.copy()
            # Continue the WHILE loop to specialize further
        # 2. Check for tie (equal accuracy but more specific)
        elif best_specialized_antecedent and best_specialized_accuracy == current_best_acc_for_comparison:
            # Found an equally accurate, but more specific rule
            if len(best_specialized_antecedent) > len(best_antecedent):
                 current_antecedent = best_specialized_antecedent
                 best_accuracy = best_specialized_accuracy
                 best_antecedent = current_antecedent.copy()
                 # Continue the WHILE loop to specialize further
            else:
                 # No strictly better rule, and the tie-breaker rule is not more specific than the current best rule.
                 break
        else:
            # No further profitable specialization. Return the best specific rule found.
            break

    # Only return a rule if it is specific (non-empty antecedent) and better than the initial IF TRUE rule.
    if best_antecedent:
        return best_antecedent, best_accuracy
    else:
        # Return None if the best found rule was the IF TRUE rule or no good rule was found
        return None, initial_accuracy


def sequential_covering_algorithm(training_data: List[Dict[str, Any]], attribute_names: List[str], target_name: str) -> Dict[str, Any]:
    """
    Implementation of the corrected Sequential Covering algorithm.
    """

    all_classes = sorted(list(set(ex[target_name] for ex in training_data)))
    learned_rules = {'all_rules': []}

    print('====================================================================')
    print('ü§ñ Starting Corrected Sequential Covering Algorithm')
    print('====================================================================')

    remaining_examples = copy.deepcopy(training_data)
    all_classes_set = set(ex[target_name] for ex in training_data)

    for target_class in all_classes:

        print(f"\n--- Learning Rules for Class: '{target_class}' ---")
        rule_count = 0

        while True:
            pos_examples_for_class = [ex for ex in remaining_examples if ex[target_name] == target_class]

            if not pos_examples_for_class:
                print(f" ¬†‚úÖ All positive examples for '{target_class}' covered.")
                break

            # Calculate the base accuracy (IF TRUE) for the stopping condition
            majority_class_count = len([ex for ex in remaining_examples if ex[target_name] == target_class])
            if len(remaining_examples) == 0: break

            base_accuracy = (majority_class_count + 1) / (len(remaining_examples) + len(all_classes_set))

            # Find the best rule
            best_antecedent, best_accuracy = find_best_rule(
                current_examples=remaining_examples,
                target_class=target_class,
                attribute_names=attribute_names,
                target_name=target_name
            )

            # Stop condition: Rule is not sufficiently accurate (must be strictly better than base) or no rule found
            if not best_antecedent or best_accuracy <= base_accuracy:
                print(f" ¬†‚ùå Stopping search for '{target_class}' (Best Acc: {best_accuracy:.4f} <= Base Acc: {base_accuracy:.4f}).")
                break

            # 4. Cover and Remove
            rule_count += 1
            learned_rules['all_rules'].append({
                'antecedent': best_antecedent,
                'consequent': target_class,
                'accuracy': best_accuracy
            })

            # Identify examples correctly covered by this new rule
            covered_and_correct = [
                ex for ex in remaining_examples
                if covers_example(best_antecedent, ex) and ex[target_name] == target_class
            ]

            # Remove the covered examples that were correctly classified
            covered_ids = set(ex.get('patient_id') for ex in covered_and_correct)
            remaining_examples = [ex for ex in remaining_examples if ex.get('patient_id') not in covered_ids]

            # For logging: get all covered examples by the rule (correct and incorrect)
            all_covered = get_covered_examples(best_antecedent, training_data) # Check against full set for total coverage count

            # Log the step
            print(f" ¬†Rule {rule_count}: {format_rule(best_antecedent, target_class, target_name)}")
            print(f" ¬† ¬†- Accuracy (Laplace): {best_accuracy:.4f}")
            print(f" ¬† ¬†- Pos Examples Covered & Removed: {len(covered_and_correct)}")
            print(f" ¬† ¬†- Examples Remaining in class: {len([ex for ex in remaining_examples if ex[target_name] == target_class])}")

    # 5. Add a default rule
    if remaining_examples:
        default_class = pd.Series([ex[target_name] for ex in remaining_examples]).mode()[0]
    else:
        default_class = pd.Series([ex[target_name] for ex in training_data]).mode()[0]

    learned_rules['default_rule'] = {'antecedent': {}, 'consequent': default_class, 'accuracy': 1.0}

    print("\n--- Final Rule Set Summary ---")
    print(f" ¬†Total Specific Rules Learned: {len(learned_rules['all_rules'])}")
    print(f" ¬†Default Rule: {format_rule({}, default_class, target_name)}")
    print('====================================================================')

    return learned_rules

# ============================================================================
#  DATA LOADING AND EXECUTION
# ============================================================================

print("üìÇ Please enter the exact file path for your 'medical_diagnosis_dataset.json' file:")
try:
    # Use standard input for the path
    # dataset_path = input("Path: ").strip()
    dataset_path = "data/medical_diagnosis_dataset.json"
except EOFError:
    # Fallback for execution environments that don't support interactive input
    print("\n‚ö†Ô∏è Input error. Using placeholder path: medical_diagnosis_dataset.json")
    dataset_path = "medical_diagnosis_dataset.json"


training_data = None
attribute_names = []
target_name = 'diagnosis'

try:
    with open(dataset_path, 'r') as file:
        dataset = json.load(file)

    training_data = dataset['training_data']
    attribute_names = list(dataset['attributes'].keys())
    df_train = pd.DataFrame(training_data) # For default rule calculation

    print(f'‚úÖ Dataset loaded successfully with {len(training_data)} examples from: {dataset_path}')

except FileNotFoundError:
    print(f"\n‚ùå Error: The file at path '{dataset_path}' was not found. Please ensure the path is correct and the file is present.")
except Exception as e:
    print(f"\n‚ùå An unexpected error occurred during data loading: {e}")


# ============================================================================
#  RUN ALGORITHM AND DISPLAY RESULTS
# ============================================================================
if training_data:
    learned_rules = sequential_covering_algorithm(training_data, attribute_names, target_name)

    print("\n\nüéØ Final Generated Classification Rules (Ordered):")
    rule_list = learned_rules['all_rules']

    for i, rule in enumerate(rule_list, 1):
        # Calculate coverage for display
        total_covered = len(get_covered_examples(rule['antecedent'], training_data))

        print(f" ¬†{i}. {format_rule(rule['antecedent'], rule['consequent'], target_name)} (Laplace Acc: {rule['accuracy']:.4f}, Covered: {total_covered})")

    default_rule = learned_rules['default_rule']
    print(f" ¬†Default. {format_rule(default_rule['antecedent'], default_rule['consequent'], target_name)} (Acc: {default_rule['accuracy']:.1f})")
````

## üéØ Section 3: FOIL Algorithm (First-Order Inductive Learner) - Relational Rule Learning

**What to Expect:**
We will implement the FOIL algorithm, a classic technique in **Inductive Logic Programming (ILP)**. Unlike the propositional rule learners (like CN2/RIPPER) that use fixed attributes and values, FOIL learns **first-order rules**. These rules contain **variables** and **predicates**, allowing them to describe **relationships** between objects.

**Why This Matters:**
First-order logic is far more expressive than propositional logic. FOIL allows us to learn complex relational concepts like *Grandparent* or *Cousin* from structural data (like the `family_relationships_dataset.json`), something traditional attribute-value learners cannot do.


### üß† The Intuition Behind FOIL

Imagine the detective (the learning algorithm) isn't just looking at the attributes of a single suspect (e.g., *tall*, *wearing a hat*), but also their relationships with others:

- **Goal:** Define the concept of `Grandparent(X, Y)`.
- **FOIL's Strategy:** Start with the most general rule: `Grandparent(X, Y) ‚Üê TRUE`
- **Rule Refinement:** Greedily add a **predicate** (a relational condition) that best distinguishes positive examples from negative ones.
    - *Iteration 1:* Add `Parent(X, Z)`. Now the rule is: `Grandparent(X, Y) ‚Üê Parent(X, Z)`.
    - *Iteration 2:* Add the linking predicate `Parent(Z, Y)`. The final rule becomes:
      $$\text{Grandparent}(\mathbf{X}, \mathbf{Y}) \leftarrow \text{Parent}(\mathbf{X}, \mathbf{Z}) \land \text{Parent}(\mathbf{Z}, \mathbf{Y})$$

FOIL is essentially a **Sequential Covering** algorithm, but instead of specializing a rule by adding attribute-value pairs, it specializes by adding **literals (predicates)** involving **variables**.


## Process Overview: FOIL Algorithm

FOIL is a greedy, top-down, sequential covering algorithm that learns one rule at a time until all positive examples are covered.

1.  **Select a Target Predicate**: Choose the relationship to learn (e.g., `Grandparent(X, Y)`).
2.  **Sequential Covering Loop (Outer Loop)**:
    a.  Initialize a new rule, $R$, to the most general form: $R: \text{Target}(X_1, \dots, X_n) \leftarrow \text{TRUE}$.
    b.  **Rule Specialization Loop (Inner Loop)**: While the rule $R$ covers negative examples:
        i.   Generate all possible new **literals** (e.g., `Parent(X, Z)`, `Male(X)`, `Sibling(Y, Z)`).
        ii.  Choose the literal that maximizes **FOIL's Information Gain** (a metric that balances covering many positive examples while excluding many negative examples).
        iii. Add the best literal to the rule $R$.
    c.  **Add Rule**: Add the learned rule $R$ to the final rule set.
    d.  **Cover and Eliminate**: Remove all **positive examples** covered by $R$ from the training set.
    e.  **Repeat**: Go back to step 2a until no positive examples remain.

**Expected Outcome:**
The output is a set of logical rules (predicates connected by $\land$) that define the target relationship. For the `Grandparent` concept, the expected outcome would be a single rule like:
$$\text{Grandparent}(\mathbf{X}, \mathbf{Y}) \leftarrow \text{Parent}(\mathbf{X}, \mathbf{Z}) \land \text{Parent}(\mathbf{Z}, \mathbf{Y})$$

````python
import json
import itertools
import math
import os
from copy import deepcopy
from typing import List, Dict, Any, Tuple, Set

# ---------------------------
# 1. Load Facts
# ---------------------------
def load_facts(facts_dict: Dict[str, List[List[str]]]) -> Dict[str, Set[Tuple]]:
    kb = {}
    for pred, tuples_list in facts_dict.items():
        if tuples_list and isinstance(tuples_list[0], str):
            kb[pred] = set((t,) for t in tuples_list)
        else:
            kb[pred] = set(tuple(t) for t in tuples_list)
    kb['Parent'] = kb.get('Father', set()).union(kb.get('Mother', set()))
    return kb

# ---------------------------
# 2. Load Dataset
# ---------------------------
def load_dataset_from_path(file_path: str, target_predicate: str) -> Tuple[Dict, List, List, List]:
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"The file '{file_path}' was not found.")
    with open(file_path, 'r') as f:
        data = json.load(f)
    kb = load_facts(data['facts'])
    people = data['people']
    target_data = data['target_concepts'][target_predicate]
    positives = target_data.get('positive_examples', [])
    negatives = target_data.get('negative_examples', [])
    return kb, people, positives, negatives

# ---------------------------
# 3. Candidate Literals
# ---------------------------
def get_all_literals(kb: Dict) -> List[Dict[str, Any]]:
    literals = []
    available_predicates = ['Parent', 'Father', 'Mother', 'Male', 'Female']
    for pred in available_predicates:
        if pred in kb:
            arity = len(next(iter(kb[pred]))) if kb[pred] else 1
            if arity in [1, 2]:
                literals.append({'name': pred, 'arity': arity})
    return literals

# ---------------------------
# 4. Format Rule
# ---------------------------
def format_rule(rule_literals: List[Dict[str, List[str]]]) -> str:
    head = rule_literals[0]
    head_str = f"{head['name']}({', '.join(head['args'])})"
    if len(rule_literals) == 1:
        return f"{head_str} ‚Üê TRUE"
    antecedents = [f"{lit['name']}({', '.join(lit['args'])})" for lit in rule_literals[1:]]
    return f"{head_str} ‚Üê {' ‚àß '.join(antecedents)}"

# ---------------------------
# 5. Fact Checking
# ---------------------------
def is_fact_true(predicate_name: str, constants: Tuple[str, ...], kb: Dict[str, Set[Tuple]]) -> bool:
    return constants in kb.get(predicate_name, set())

# ---------------------------
# 6. Find Variable Bindings
# ---------------------------
def find_bindings(rule_antecedent: List[Dict[str, List[str]]], example: List[str], kb: Dict[str, Set[Tuple]], people: List[str]) -> List[Dict[str, str]]:
    if not rule_antecedent:
        return [{'X': example[0], 'Y': example[1]}]
    initial_binding = {'X': example[0], 'Y': example[1]}
    current_bindings: List[Dict[str, str]] = [initial_binding]
    for literal in rule_antecedent:
        next_bindings = []
        lit_vars = literal['args']
        pred_name = literal['name']
        for binding in current_bindings:
            unbound_indices = [i for i, v in enumerate(lit_vars) if v not in binding]
            if not unbound_indices:
                constants = tuple(binding[v] for v in lit_vars)
                if is_fact_true(pred_name, constants, kb):
                    next_bindings.append(binding)
                continue
            for new_constants in itertools.product(people, repeat=len(unbound_indices)):
                temp_binding = binding.copy()
                for i, new_c in zip(unbound_indices, new_constants):
                    temp_binding[lit_vars[i]] = new_c
                literal_constants = tuple(temp_binding[v] for v in lit_vars)
                if is_fact_true(pred_name, literal_constants, kb):
                    next_bindings.append(temp_binding)
        current_bindings = next_bindings
        if not current_bindings:
            return []
    return current_bindings

# ---------------------------
# 7. FOIL Gain Calculation
# ---------------------------
def calculate_foil_gain(P0: int, N0: int, P1: int, N1: int) -> float:
    if P1 == 0 or (P1 + N1) == 0:
        return 0.0
    I_R = math.log2(P0/(P0 + N0)) if P0 > 0 else 0.0
    I_R_L = math.log2(P1/(P1 + N1))
    return P1 * (I_R_L - I_R)
````

````python
# ============================================================================
# FOIL ALGORITHM IMPLEMENTATION - FIRST-ORDER INDUCTIVE LEARNER
# ============================================================================

def foil_algorithm(target_name: str, positive_examples: List[List[str]], negative_examples: List[List[str]], kb: Dict[str, Set[Tuple]], people: List[str]):
    """
    Implements the FOIL (First-Order Inductive Learner) algorithm for learning first-order logic rules.
    
    Example: Learning Grandparent(X, Y) from family relationships
    Input: target_name = "Grandparent"
           positive_examples = [["George", "Tom"], ["Mary", "Alice"], ...]
           negative_examples = [["George", "John"], ["Tom", "Frank"], ...]
           kb = knowledge base with Father, Mother, Parent facts
           people = ["George", "Mary", "John", "Tom", ...]
    
    Output: Rules like Grandparent(X, Y) ‚Üê Parent(X, Z) ‚àß Parent(Z, Y)
    """
    
    # Initialize storage for learned rules and coverage statistics
    learned_rules = []  # Will store the final set of rules learned
    coverage_data = []  # Will track how many examples each rule covers (for visualization)
    
    # Create a working copy of positive examples that will be reduced as rules cover them
    # Example: remaining_positives starts as [["George", "Tom"], ["Mary", "Alice"], ...]
    remaining_positives = deepcopy(positive_examples)
    
    # Get all available predicates from the knowledge base (Father, Mother, Parent, Male, Female)
    all_literals = get_all_literals(kb)
    
    # Counter for rule numbering (Rule 1, Rule 2, etc.)
    rule_id = 0

    # ========================================================================
    # SEQUENTIAL COVERING LOOP - Learn one rule at a time until all positives are covered
    # ========================================================================
    while remaining_positives:  # Continue until no positive examples remain uncovered
        # Start learning a new rule
        rule_id += 1
        
        # Initialize rule with just the head: Grandparent(X, Y) ‚Üê TRUE
        # This is the most general possible rule for the target concept
        current_rule = [{'name': target_name, 'args': ['X', 'Y']}]
        
        # Count current positive and negative examples for FOIL gain calculation
        # P0 = number of remaining positive examples (e.g., 10 grandparent pairs left)
        # N0 = total negative examples (stays constant, e.g., 8 non-grandparent pairs)
        P0, N0 = len(remaining_positives), len(negative_examples)

        # ====================================================================
        # RULE SPECIALIZATION LOOP - Add literals to make the rule more specific
        # ====================================================================
        while True:  # Keep adding literals until no improvement or rule is pure
            # Initialize variables for finding the best literal to add
            best_gain = -1      # Best FOIL gain found so far
            best_literal = None # The literal that gives the best gain
            
            # Get all variables currently used in the rule (initially just X, Y)
            # Example: after adding Parent(X, Z), current_vars = {'X', 'Y', 'Z'}
            current_vars = set().union(*[lit['args'] for lit in current_rule])
            
            # Storage for all possible new literals we could add
            candidate_literals = []

            # ================================================================
            # CANDIDATE LITERAL GENERATION
            # ================================================================
            # Generate all possible literals that could be added to the current rule
            for lit_info in all_literals:
                pred, arity = lit_info['name'], lit_info['arity']  # e.g., 'Parent', 2 (arity = number of arguments the predicate takes)

                # Create potential variable combinations including existing vars + new var Z
                # Example: if current_vars = {'X', 'Y'}, potential_args = ['X', 'Y', 'Z']
                potential_args = list(current_vars.union({'Z'}))
                
                # Try all possible argument combinations for this predicate
                # Example: For Parent (arity=2), try Parent(X,Y), Parent(X,Z), Parent(Y,Z), etc.
                for args_combo in itertools.product(potential_args, repeat=arity):
                    
                    # Skip if this exact literal is already in the rule
                    # Example: Don't add Parent(X,Z) if it's already there
                    if any(l['name']==pred and l['args']==list(args_combo) for l in current_rule[1:]):
                        continue
                    
                    # CONSTRAINT: Don't introduce more than one new variable at once
                    # Example: If current_vars = {'X', 'Y'}, don't allow Parent(Z, W)
                    if sum(1 for v in args_combo if v not in current_vars) > 1:
                        continue
                    
                    # SPECIAL CONSTRAINT for Grandparent learning:
                    # Only allow Parent/Father/Mother with specific argument patterns
                    # This guides the search toward meaningful grandparent relationships
                    if target_name.lower() == 'grandparent' and pred in ['Parent','Father','Mother']:
                        # Only allow Parent(X,Z) or Parent(Z,Y) - the chain X‚ÜíZ‚ÜíY
                        if set(args_combo) not in [{'X','Z'}, {'Z','Y'}]:
                            continue
                    
                    # Add this literal as a valid candidate
                    candidate_literals.append({'name': pred, 'args': list(args_combo)})

            # If no valid candidates found, stop specializing this rule
            if not candidate_literals: 
                break

            # ================================================================
            # EVALUATE EACH CANDIDATE LITERAL
            # ================================================================
            # Get the current rule body (everything after the head)
            # Example: If current_rule = [Grandparent(X,Y), Parent(X,Z)], 
            #          then current_rule_antecedent = [Parent(X,Z)]
            current_rule_antecedent = current_rule[1:]
            
            # Test each candidate literal to see which gives the best FOIL gain
            for literal in candidate_literals:
                # Create new rule by adding this literal
                # Example: [Parent(X,Z)] + [Parent(Z,Y)] = [Parent(X,Z), Parent(Z,Y)]
                new_rule_antecedent = current_rule_antecedent + [literal]
                
                # Count how many positive examples this new rule covers
                # Example: How many remaining grandparent pairs satisfy Parent(X,Z) ‚àß Parent(Z,Y)?
                P1_count = sum(1 for ex in remaining_positives if find_bindings(new_rule_antecedent, ex, kb, people))
                
                # Count how many negative examples this new rule covers (we want this low)
                # Example: How many non-grandparent pairs satisfy Parent(X,Z) ‚àß Parent(Z,Y)?
                N1_count = sum(1 for ex in negative_examples if find_bindings(new_rule_antecedent, ex, kb, people))
                
                # Calculate FOIL gain for this literal
                # FOIL gain = P1 * (log2(P1/(P1+N1)) - log2(P0/(P0+N0)))
                # Higher gain = better literal (covers more positives, fewer negatives)
                gain = calculate_foil_gain(P0, N0, P1_count, N1_count)
                
                # Keep track of the best literal found so far
                if gain > best_gain:
                    best_gain, best_literal, best_P1, best_N1 = gain, literal, P1_count, N1_count

            # ================================================================
            # DECIDE WHETHER TO ADD THE BEST LITERAL
            # ================================================================
            if best_gain > 0:  # Only add literal if it improves the rule
                # Add the best literal to our rule
                # Example: Grandparent(X,Y) ‚Üê Parent(X,Z) becomes 
                #          Grandparent(X,Y) ‚Üê Parent(X,Z) ‚àß Parent(Z,Y)
                current_rule.append(best_literal)
                
                # Update counts for next iteration
                P0, N0 = best_P1, best_N1
                
                # If rule covers no negative examples, it's perfect - stop specializing
                if N0 == 0: 
                    break
            else:
                # No literal improves the rule - stop specializing
                break

        # ====================================================================
        # COVER AND REMOVE EXAMPLES
        # ====================================================================
        # Find all positive examples that this learned rule covers correctly
        covered_indices = []
        current_rule_antecedent = current_rule[1:]  # Get rule body (without head)
        
        # Check each remaining positive example
        for i, ex in enumerate(remaining_positives):
            # If the rule's conditions are satisfied by this example
            # Example: Does ["George", "Tom"] satisfy Parent(X,Z) ‚àß Parent(Z,Y)?
            if find_bindings(current_rule_antecedent, ex, kb, people):
                covered_indices.append(i)  # Mark this example as covered
        
        # Remove covered examples from the remaining set (sequential covering)
        covered_count = 0
        for index in sorted(covered_indices, reverse=True):  # Remove from end to avoid index shifts
            remaining_positives.pop(index)
            covered_count += 1

        # ====================================================================
        # STORE THE LEARNED RULE
        # ====================================================================
        # Only store rules that have actual conditions and cover at least one example
        if len(current_rule) > 1 and covered_count > 0:
            learned_rules.append(current_rule)
            
            # Store coverage data for visualization
            # Example: "Rule 1 covers 6 examples with conditions Mother(X,Z) ‚àß Parent(Z,Y)"
            coverage_data.append({
                'Rule_ID': f"Rule {rule_id}",
                'Examples_Covered': covered_count,
                'Rule_Antecedent': format_rule(current_rule).split('‚Üê')[1].strip()  # Get just the conditions part
            })

    # Return both the learned rules and coverage statistics
    return learned_rules, coverage_data
````

````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def visualize_foil_coverage_dynamic(coverage_data: List[Dict[str, Any]], target_name: str):
    if not coverage_data: return
    df = pd.DataFrame(coverage_data)
    total_positives = df['Examples_Covered'].sum()
    fig, ax = plt.subplots(figsize=(12,7))
    bars = ax.bar(df['Rule_ID'], df['Examples_Covered'], color='#5cb85c', alpha=0.8)
    ax.set_title(f'FOIL Algorithm: Coverage of "{target_name}" Concept (Total Positives: {total_positives})', fontsize=15, fontweight='bold')
    ax.set_xlabel('Learned Rule', fontsize=12)
    ax.set_ylabel('Number of Positive Examples Covered', fontsize=12)
    ax.set_yticks(np.arange(0, df['Examples_Covered'].max()+1, 1))
    ax.grid(axis='y', linestyle='--', alpha=0.5)
    for i, bar in enumerate(bars):
        height = bar.get_height()
        ax.text(bar.get_x()+bar.get_width()/2., height+0.1, f'{int(height)}', ha='center', va='bottom', fontsize=11, fontweight='bold')
        rule_text = f"R{i+1}: ‚Üê {df['Rule_Antecedent'].iloc[i]}"
        ax.text(bar.get_x()+bar.get_width()/2., -0.7-(i%2)*0.4, rule_text.replace('‚àß','\n‚àß'), ha='center', va='top', fontsize=9, rotation=25, color='#333333')
    plt.subplots_adjust(bottom=0.35)
    plt.show()
````

## ‚ùì Frequently Asked Questions (FAQ) for the FOIL Experiment

### Technical Questions (FOIL Core)

**Q1: What is the fundamental difference between FOIL and propositional algorithms (like Find-S or ID3)?**

A: The main difference lies in the **Hypothesis Representation**.
* **FOIL** learns **First-Order Rules** (using **predicates** and **variables**). It can capture **relationships** (e.g., $\text{Grandparent}(\mathbf{X}, \mathbf{Y}) \leftarrow \text{Parent}(\mathbf{X}, \mathbf{Z}) \land \text{Parent}(\mathbf{Z}, \mathbf{Y})$).
* **Propositional Algorithms** learn rules based only on **fixed attributes** of a single entity. FOIL is necessary for relational learning.

**Q2: What is FOIL Gain and how does it drive the algorithm?**

A: **FOIL Gain** is the metric used to guide the **greedy search** during rule specialization. It's a derivative of Information Gain that measures the value of adding a new literal ($L$) to the current rule ($R$) by prioritizing literals that maximize the coverage of **remaining positive examples ($P_1$)** while minimizing the coverage of **negative examples ($N_1$)**.

**Q3: How does FOIL handle new variables, like 'Z' in your rules?**

A: The introduction of a new variable (like $Z$) is the mechanism that allows FOIL to discover **relational chains**. A new variable must always be **"linked"** to an existing variable ($X$ or $Y$) by the new literal, allowing the rule to describe multi-step relationships (e.g., grandparent-parent-child).

### Questions on FOIL Output (The Grandparent Rules)

**Q4: Why did the algorithm generate six complex and redundant rules instead of the single, simple rule: $\text{Grandparent}(\mathbf{X}, \mathbf{Y}) \leftarrow \text{Parent}(\mathbf{X}, \mathbf{Z}) \land \text{Parent}(\mathbf{Z}, \mathbf{Y})$?**

A: This is a demonstration of the **greedy search limitation** and **sequential covering bias**.
1.  **Early Specificity:** The first rule found a high-gain, *specific* rule (`Mother(X, Z) ‚àß Parent(Z, Y)`), which was accepted.
2.  **Sequential Removal:** By removing the examples covered, the remaining pool of positives changed, forcing the subsequent greedy search to find highly specific **"patch rules"** (Rules 2, 3, 5, 6) for the leftovers instead of the underlying general concept.

**Q5: Rule 3 was learned as $\text{Grandparent}(\mathbf{X}, \mathbf{Y}) \leftarrow \text{Female}(\mathbf{Y})$. Why was this rule, which covered 2 negative examples (P/N: 4/2), accepted?**

A: The rule was accepted because the **specialization loop** for Rule 3 stalled: **no subsequent literal** could be found that yielded a **positive FOIL Gain** to further refine the rule and remove the 2 negatives. This illustrates the trade-off between maximizing coverage (4 positives) and achieving perfect purity when the greedy search fails to find an immediate improvement.

**Q6: What is the key limitation of FOIL in real-world scenarios?**

A: **Noise and Error Sensitivity.** FOIL assumes perfect, noise-free data. A single mislabeled example can drastically complicate the learning process, forcing the generation of highly complex, low-coverage rules (overfitting) just to "patch" that one outlier.

### Key Technical Concepts

**1. Inductive Logic Programming (ILP):** A subfield of ML that uses first-order logic (FOL) to represent data and hypotheses, essential for learning **relationships** rather than just attributes.

**2. Sequential Covering:** A general strategy used by FOIL where the algorithm learns a single good rule, removes the examples it covers, and repeats until no positive examples remain. The final concept is a disjunction (OR) of all learned rules.

**3. FOIL's Inductive Bias:** The algorithm has a strong bias toward learning **pure** (zero negative examples) and **highly specialized** rules due to its greedy nature. This often leads to fragmented, over-specific rules rather than a single general one.
