## üìñ Background & Theory

### What is a Self-Organizing Map (SOM)?

A **Self-Organizing Map** (also called **Kohonen Map**) is:
- A **neural network** trained using **unsupervised learning**
- Used to represent **high-dimensional data** on a **low-dimensional (usually 2D)** grid
- Preserves **topological properties** ‚Äî meaning similar input data map to nearby neurons

Invented by **Teuvo Kohonen (1982)**.

### Key Concepts

**1. Self-Organizing Map (SOM)**
- Type of artificial neural network
- Unsupervised learning algorithm
- Creates low-dimensional (usually 2D) representation of high-dimensional data
- Preserves topological properties of input space
- Invented by Teuvo Kohonen in 1982 (also called Kohonen Map)  

1524859998652.png

**2. Competitive Learning**
- Each neuron competes to respond to the input.
- The **winner** (Best Matching Unit or BMU) is the neuron closest to the input vector.
- Only the BMU and its **neighbors** are updated.

**3. Topology Preservation**
- Similar data ‚Üí mapped to nearby neurons.
- Dissimilar data ‚Üí mapped farther apart.
- Ensures smooth transitions between different patterns.

### Mathematical Foundation

**SOM Structure:**
```
- Input Layer: n-dimensional input vector x = [x‚ÇÅ, x‚ÇÇ, ..., x‚Çô]
- Output Layer: 2D grid of neurons (e.g., 10√ó10 = 100 neurons)
- Weight Vectors: Each neuron has weight vector w = [w‚ÇÅ, w‚ÇÇ, ..., w‚Çô]
```

## üß© Training Algorithm (Simple Explanation)

Here‚Äôs how a Self-Organizing Map (SOM) learns ‚Äî step by step üëá

### Step 1Ô∏è‚É£: Initialize
- We start with a **grid of neurons** (for example, a 10√ó10 grid = 100 neurons).  
- Each neuron has some **weights** (like small ‚Äúprofiles‚Äù) with random numbers at the start.  
- These weights will slowly adjust to represent real data patterns.

---

### Step 2Ô∏è‚É£: Show the Data (Input Vector)
- Take **one data point** from your dataset ‚Äî for example,  
  $(x_1, x_2, x_3, ... , x_n)$  
  (it could be something like `[Age, Income, Spending]`).
- This data point is shown to all the neurons.

---

### Step 3Ô∏è‚É£: Find the Best Matching Neuron (BMU)
- Each neuron checks **how close it is** to the input data.
- The distance between the input and a neuron‚Äôs weight vector is calculated (usually Euclidean distance):

  $$
  \text{distance} = \sqrt{(x_1 - w_1)^2 + (x_2 - w_2)^2 + ... + (x_n - w_n)^2}
  $$

- The neuron with the **smallest distance** wins!  
  This neuron is called the **Best Matching Unit (BMU)** ‚Äî it‚Äôs the one most similar to the current input.

---

### Step 4Ô∏è‚É£: Update the BMU and Its Neighbors
- The BMU and its nearby neurons **move a bit closer** to the input vector.  
- Neurons closer to the BMU move **more**, while far ones move **less**.

  Simplified formula:
  $$
  \text{new weight} = \text{old weight} + \text{learning rate} \times (\text{input} - \text{old weight})
  $$

- This makes the map ‚Äúpull‚Äù toward the input pattern gradually.

---

### Step 5Ô∏è‚É£: Reduce Learning Over Time
- In the beginning, neurons move a lot (high learning rate).
- As training continues, we reduce:
  - **Learning rate** (how much weights change)
  - **Neighborhood size** (how many nearby neurons are affected)

  This helps the map stabilize over time.

---

### Step 6Ô∏è‚É£: Repeat Until the Map Learns
- We repeat steps 2‚Äì5 for many rounds (called **epochs**).
- Each round, the map becomes more organized ‚Äî neurons learn to represent different groups of data.

---

‚úÖ **Result:**  
After training, nearby neurons on the SOM represent **similar data points**, and distant neurons represent **different kinds of data**.  
This gives you a beautiful **2D map of complex high-dimensional patterns**.


### SOM vs Other Clustering Methods

| Aspect | SOM | K-Means | Hierarchical |
|--------|-----|---------|-------------|
| **Topology** | Preserved | Not preserved | Preserved (tree) |
| **Visualization** | 2D map | Scatter plots | Dendrogram |
| **Clusters** | Soft boundaries | Hard boundaries | Hierarchical |
| **Dimensionality** | Reduces to 2D | Same as input | Same as input |
| **Interpretability** | Excellent | Good | Good |
| **Speed** | Moderate | Fast | Slow |

**Key Advantage of SOM**: Creates interpretable 2D visualization while preserving data topology!

# üß† Self-Organizing Map (SOM) ‚Äî Step-by-Step Training Process

## üå± What We Start With

Before training begins, we have **two main things:**

1. **The Dataset (Input Vectors)**  
   This is your collection of data points.  
   Example dataset:  
| Age | Income  |
|-----|----------|
| 20  | 30,000  |
| 35  | 50,000  |
| 60  | 90,000  |

Each row can be written as a vector ‚Üí  
$(20, 30000)$, $(35, 50000)$, $(60, 90000)$

2. **The SOM Grid (Neurons / Nodes)**  
- Imagine a **2D grid** of small circles (neurons).  
- Each neuron has a **weight vector**, which has the same number of components as your input data.  
  For example, $(w_{age}, w_{income})$.  
- Initially, these weights are **random numbers** ‚Äî they don‚Äôt yet represent anything meaningful.

---

## ‚öôÔ∏è Step 1: Input Vector is Presented

We begin training by selecting **one data point (input vector)** from the dataset.  
For example, let's choose:
> $(35, 50000)$

This vector will be compared with every neuron on the grid.

üí° Think of it like:
> ‚ÄúLet‚Äôs show this customer (age 35, income 50,000) to all the neurons ‚Äî and see which neuron already looks the most like this customer.‚Äù

---

## üèÜ Step 2: Competitive Step (Finding the BMU)

Each neuron measures **how close it is** to the input vector using **Euclidean distance**:

$$
distance = \sqrt{(x_1 - w_1)^2 + (x_2 - w_2)^2}
$$

The neuron with the **smallest distance** is called the **Best Matching Unit (BMU)**.

Example:

| Neuron | Weight (Age, Income) | Distance to (35, 50000) |
|:-------:|:--------------------:|:------------------------:|
| A | (25, 25000) | Large |
| B | (33, 52000) | **Smallest ‚úÖ** |
| C | (55, 80000) | Big |
| D | (40, 70000) | Medium |

Neuron **B** is the **BMU** because it is closest to the input.

---

## ü§ù Step 3: Cooperative Step (Neighborhood Update)

Once the BMU is identified:

1. The BMU **updates its weight vector** to move closer to the input vector:
$$
w_{\text{new}} = w_{\text{old}} + \eta (x - w_{\text{old}})
$$
where $\eta$ (eta) is the learning rate (e.g., 0.1)

2. **Nearby neurons** (around the BMU on the grid) also move toward the input,  
but **less strongly** depending on their distance from the BMU.  
This effect is called the **neighborhood function**.

Example:
- BMU (B) moves a lot toward $(35, 50000)$  
- Neighbor (D) moves a little  
- Far neuron (A) barely moves or doesn‚Äôt move at all

---

## üîÅ Step 4: Repeat for All Inputs

- Pick the next input vector (say $(20, 30000)$), and repeat the same three steps:
1. Find BMU  
2. Update BMU and its neighbors  

Do this **for all data points**, and repeat for multiple training rounds.

Over time:
- The SOM grid **self-organizes**.
- Neurons representing similar kinds of inputs **cluster together** on the map.

---

## üéØ Real-World Analogy

Imagine each neuron as a **shopkeeper** trying to match the taste of customers.

1. A customer walks in (input vector).  
2. Each shopkeeper compares their current style with the customer‚Äôs taste.  
3. The one who matches best (BMU) learns the most about that customer.  
4. Nearby shopkeepers also learn a bit (neighbors update).  
5. After many customers, nearby shops start offering similar things ‚Äî forming clusters like:
- ‚ÄúTech lovers‚Äù
- ‚ÄúBudget buyers‚Äù
- ‚ÄúLuxury customers‚Äù

---

‚ú® **In short:**
1. Present an input vector  
2. Find the BMU  
3. Update BMU + neighbors  
4. Repeat for all data  

Over time, the SOM organizes itself ‚Äî forming meaningful clusters based on data similarity.

## üîß Section 4: Data Preprocessing

### What to Expect
We'll prepare the data for SOM by selecting features and normalizing them.

### Process Overview
1. Select features for SOM (exclude PersonID)
2. Normalize features to [0, 1] range
3. Prepare data matrix for SOM training

### Expected Outcome
Normalized feature matrix ready for SOM.

### Why This Matters
SOM works best with normalized data where all features contribute equally.

````python
# Select features for SOM (exclude PersonID)
som_features = ['Red', 'Green', 'Blue', 'Extraversion', 'Creativity', 'Calmness']

# Extract feature matrix
X = df[som_features].values  # Convert to NumPy array

print("="*80)
print("üéØ FEATURE SELECTION")
print("="*80)
print(f"Selected Features: {', '.join(som_features)}")
print(f"Feature Matrix Shape: {X.shape}")
print(f"  ‚Üí {X.shape[0]} people (samples)")
print(f"  ‚Üí {X.shape[1]} features (dimensions)")

# Display original feature ranges
print("\n" + "="*80)
print("üìä ORIGINAL FEATURE RANGES (Before Normalization)")
print("="*80)
for idx, feature in enumerate(som_features):  # Loop through features
    min_val = X[:, idx].min()  # Minimum value
    max_val = X[:, idx].max()  # Maximum value
    mean_val = X[:, idx].mean()  # Mean value
    print(f"{feature:15s}: Min={min_val:6.1f}, Max={max_val:6.1f}, Mean={mean_val:6.1f}")

# Normalize data to [0, 1] range
# SOM typically works better with normalized data
X_normalized = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))  # Min-max normalization

print("\n" + "="*80)
print("‚öôÔ∏è  NORMALIZATION APPLIED")
print("="*80)
print("Normalization Method: Min-Max Scaling to [0, 1]")
print("Formula: X_norm = (X - X_min) / (X_max - X_min)")

# Display normalized feature ranges
print("\n" + "="*80)
print("üìä NORMALIZED FEATURE RANGES (After Normalization)")
print("="*80)
for idx, feature in enumerate(som_features):  # Loop through features
    min_val = X_normalized[:, idx].min()  # Should be 0
    max_val = X_normalized[:, idx].max()  # Should be 1
    mean_val = X_normalized[:, idx].mean()  # Mean value
    print(f"{feature:15s}: Min={min_val:6.3f}, Max={max_val:6.3f}, Mean={mean_val:6.3f}")

print("\n‚úÖ All features normalized to [0, 1] range!")
print("üí° Each feature now contributes equally to SOM training")
print("üéØ Data is ready for Self-Organizing Map!")
````

````python
# Configure SOM parameters
print("="*80)
print("‚öôÔ∏è  SOM CONFIGURATION")
print("="*80)

# Grid dimensions
som_x = 10  # Width of SOM grid
som_y = 10  # Height of SOM grid
print(f"Grid Size: {som_x} √ó {som_y} = {som_x * som_y} neurons")
print(f"  ‚Üí Each neuron represents a 'prototype' personality-color profile")

# Input dimensions
input_len = X_normalized.shape[1]  # Number of features (6)
print(f"\nInput Dimensions: {input_len} features")
print(f"  ‚Üí {', '.join(som_features)}")

# Training parameters
num_iterations = 10000  # Number of training iterations
initial_learning_rate = 0.5  # Starting learning rate
initial_sigma = 3.0  # Starting neighborhood radius

print(f"\nTraining Parameters:")
print(f"  ‚Ä¢ Iterations: {num_iterations:,}")
print(f"  ‚Ä¢ Initial Learning Rate: {initial_learning_rate}")
print(f"  ‚Ä¢ Initial Neighborhood Radius (œÉ): {initial_sigma}")
print(f"  ‚Ä¢ Initialization: Random")
print(f"  ‚Ä¢ Neighborhood Function: Gaussian")

# Initialize SOM
print("\n" + "="*80)
print("üé≤ INITIALIZING SOM")
print("="*80)

som = MiniSom(x=som_x,  # Grid width
             y=som_y,  # Grid height
             input_len=input_len,  # Number of input features
             sigma=initial_sigma,  # Initial neighborhood radius
             learning_rate=initial_learning_rate,  # Initial learning rate
             random_seed=42)  # For reproducibility

# Initialize weights randomly
som.random_weights_init(X_normalized)  # Initialize with data range

print("‚úÖ SOM initialized successfully!")
print(f"   Weight matrix shape: {som_x} √ó {som_y} √ó {input_len}")
print(f"   Total weights: {som_x * som_y * input_len:,}")

# Train SOM
print("\n" + "="*80)
print("üèãÔ∏è  TRAINING SOM (This may take a few seconds...)")
print("="*80)

import time
start_time = time.time()  # Record start time

# Train with random sampling
som.train_random(data=X_normalized,  # Training data
                num_iteration=num_iterations,  # Number of iterations
                verbose=False)  # Don't print progress

end_time = time.time()  # Record end time
training_time = end_time - start_time  # Calculate duration

print(f"‚úÖ Training complete!")
print(f"   Training time: {training_time:.2f} seconds")
print(f"   Iterations: {num_iterations:,}")
print(f"   Samples per iteration: 1 (random sampling)")
print(f"   Total updates: {num_iterations:,}")

# Calculate quantization error
quantization_error = som.quantization_error(X_normalized)  # Average distance to BMU
print(f"\nüìä Quantization Error: {quantization_error:.4f}")
print(f"   ‚Üí Average distance from each sample to its Best Matching Unit (BMU)")
print(f"   ‚Üí Lower is better (indicates tighter clustering)")

# Calculate topographic error
topographic_error = som.topographic_error(X_normalized)  # Topology preservation
print(f"\nüìä Topographic Error: {topographic_error:.4f}")
print(f"   ‚Üí Proportion of samples where 1st and 2nd BMUs are not adjacent")
print(f"   ‚Üí Lower is better (indicates better topology preservation)")

print("\n" + "="*80)
print("üéâ SOM TRAINING SUCCESSFUL!")
print("="*80)
print("The SOM has organized 100 people into a 10√ó10 grid based on similarity!")
print("Similar personality-color profiles are now placed close together.")
print("\nüéØ Ready to visualize the results!")
````

## ‚ùì Frequently Asked Questions (FAQs)

### 1. What is a Self-Organizing Map (SOM)?

**Answer**: A Self-Organizing Map is a type of artificial neural network that uses unsupervised learning to create a low-dimensional (typically 2D) representation of high-dimensional data while preserving topological relationships. It was invented by Teuvo Kohonen in 1982.

**Key Features**:
- Unsupervised learning (no labels needed)
- Topology preservation (similar inputs map to nearby neurons)
- Dimensionality reduction (6D ‚Üí 2D in our case)
- Competitive learning (neurons compete to respond)

---

### 2. How does SOM differ from K-Means clustering?

**Answer**: While both are unsupervised learning methods, they have key differences:

| Aspect | SOM | K-Means |
|--------|-----|--------|
| **Output** | 2D topology-preserving map | Cluster labels |
| **Topology** | Preserved (smooth transitions) | Not preserved |
| **Visualization** | Built-in 2D visualization | Requires separate plotting |
| **Boundaries** | Soft, gradual transitions | Hard cluster boundaries |
| **Interpretability** | Excellent (visual map) | Good (cluster statistics) |
| **Use Case** | Data exploration, visualization | Segmentation, classification |

**When to use SOM**: When you want to visualize high-dimensional data and discover smooth patterns.  
**When to use K-Means**: When you need clear cluster assignments for segmentation.

---

### 3. How do I choose the SOM grid size?

**Answer**: Grid size affects the granularity of the map:

**Rule of Thumb**: Grid size ‚âà 5‚àön, where n = number of samples
- For 100 samples: 5‚àö100 = 50 neurons ‚Üí 7√ó7 or 10√ó10 grid
- For 1000 samples: 5‚àö1000 ‚âà 158 neurons ‚Üí 12√ó13 or 15√ó11 grid

**Trade-offs**:
- **Small grid** (5√ó5): Coarse representation, faster training, less detail
- **Medium grid** (10√ó10): Balanced, good for most applications
- **Large grid** (20√ó20): Fine detail, slower training, may overfit

**Best Practice**: Start with 10√ó10, then experiment with different sizes.

---

### 4. What is the U-Matrix and how do I interpret it?

**Answer**: The U-Matrix (Unified Distance Matrix) visualizes cluster structure:

**What it shows**: Average distance between each neuron and its neighbors

**Interpretation**:
- **Light regions** (low distance) = Cluster centers (homogeneous areas)
- **Dark regions** (high distance) = Cluster boundaries (transitions)
- **Valleys** = Dense clusters
- **Ridges** = Separations between clusters

**How to use it**: Look for distinct light regions separated by dark boundaries to identify natural clusters.

---

### 5. What are component planes?

**Answer**: Component planes show how individual features are distributed across the SOM.

**One plane per feature**: In our case, 6 component planes (Red, Green, Blue, Extraversion, Creativity, Calmness)

**Interpretation**:
- **Red/Hot regions** = High values for that feature
- **Blue/Cold regions** = Low values for that feature

**Use cases**:
- Identify where specific characteristics are located
- Compare planes to find correlations (e.g., Red preference vs Extraversion)
- Understand what each cluster represents

**Example**: If the Red component plane is hot in the top-left corner, that region represents "red lovers."

---

### 6. What is quantization error?

**Answer**: Quantization error measures how well the SOM represents the data.

**Definition**: Average distance from each data point to its Best Matching Unit (BMU)

**Formula**: QE = (1/N) √ó Œ£ ||x·µ¢ - w_BMU(i)||

**Interpretation**:
- **Lower is better** (tighter fit to data)
- Typical range: 0.1 - 0.5 for normalized data
- Very low (<0.05) may indicate overfitting
- Very high (>1.0) may indicate underfitting

**Use case**: Compare different SOM configurations to find the best fit.

---

### 7. What is topographic error?

**Answer**: Topographic error measures how well the SOM preserves topology.

**Definition**: Proportion of data points where the 1st and 2nd BMUs are not adjacent

**Interpretation**:
- **Lower is better** (better topology preservation)
- Range: 0.0 (perfect) to 1.0 (poor)
- Typical good value: < 0.1

**Why it matters**: Low topographic error means the SOM creates smooth, continuous transitions between different regions.

---

### 8. How long should I train the SOM?

**Answer**: Training iterations depend on data size and grid size.

**Rule of Thumb**: 100-500 iterations per neuron
- For 10√ó10 grid (100 neurons): 10,000 - 50,000 iterations
- For 20√ó20 grid (400 neurons): 40,000 - 200,000 iterations

**In practice**:
- Start with 10,000 iterations
- Monitor quantization error
- Stop when error plateaus

**Training time**: For 100 samples, 10,000 iterations takes 1-2 seconds on modern hardware.

### 9. Can SOM handle categorical features?

**Answer**: Not directly, but you can encode them.

**Problem**: SOM uses Euclidean distance, which requires numerical features.

**Solutions**:
1. **One-Hot Encoding**: Convert categories to binary features
   - Example: Color = {Red, Green, Blue} ‚Üí [1,0,0], [0,1,0], [0,0,1]
2. **Ordinal Encoding**: For ordered categories (Low, Medium, High ‚Üí 0, 1, 2)
3. **Target Encoding**: Replace category with mean of target variable

**Best Practice**: Use one-hot encoding for nominal categories, ordinal encoding for ordered categories.

### 11.1 Growing Self-Organizing Map (GSOM)

**Concept**: GSOM starts with a small grid and dynamically adds neurons where needed based on data density.

**Advantages**:
- No need to specify grid size upfront
- Adapts to data complexity
- More efficient for sparse data
- Better representation of data distribution

**How it works**:
1. Start with small grid (e.g., 2√ó2)
2. Train with standard SOM algorithm
3. When a neuron's error exceeds threshold, add new neurons around it
4. Repeat until convergence or max size reached

````python
# Simplified Growing SOM Implementation (Conceptual)
print("="*80)
print("üå± GROWING SELF-ORGANIZING MAP (GSOM) - CONCEPTUAL IMPLEMENTATION")
print("="*80)

class SimpleGrowingSOM:
    """Simplified Growing SOM for educational purposes"""

    def __init__(self, initial_size=3, max_size=15, growth_threshold=0.3, input_len=6):
        """
        Initialize Growing SOM

        Parameters:
        - initial_size: Starting grid size (e.g., 3 means 3√ó3 grid)
        - max_size: Maximum grid size
        - growth_threshold: Error threshold for adding neurons
        - input_len: Number of input features
        """
        self.current_size = initial_size  # Current grid size
        self.max_size = max_size  # Maximum allowed size
        self.growth_threshold = growth_threshold  # Threshold for growth
        self.input_len = input_len  # Number of features
        self.growth_history = []  # Track growth over time

        print(f"Initialized GSOM:")
        print(f"  ‚Ä¢ Initial grid size: {initial_size}√ó{initial_size} = {initial_size**2} neurons")
        print(f"  ‚Ä¢ Maximum grid size: {max_size}√ó{max_size} = {max_size**2} neurons")
        print(f"  ‚Ä¢ Growth threshold: {growth_threshold}")
        print(f"  ‚Ä¢ Input dimensions: {input_len}")

    def simulate_growth(self, data, num_iterations=5):
        """
        Simulate GSOM growth process

        In a real implementation, this would:
        1. Train SOM on current grid
        2. Calculate quantization error for each neuron
        3. Add neurons where error > threshold
        4. Repeat until convergence
        """
        print(f"\nüå± Simulating GSOM growth over {num_iterations} iterations...\n")

        for iteration in range(num_iterations):
            # Simulate training and error calculation
            # In real implementation, would train SOM and calculate actual errors

            # Simulate growth decision (simplified)
            if self.current_size < self.max_size and iteration % 2 == 1:
                # Grow the grid
                old_size = self.current_size
                self.current_size += 1  # Increase grid size
                new_neurons = self.current_size**2 - old_size**2  # Calculate new neurons

                print(f"Iteration {iteration + 1}:")
                print(f"  ‚úÖ Grid grew from {old_size}√ó{old_size} to {self.current_size}√ó{self.current_size}")
                print(f"  ‚ûï Added {new_neurons} new neurons")
                print(f"  üìä Total neurons: {self.current_size**2}")

                self.growth_history.append({
                    'iteration': iteration + 1,
                    'grid_size': self.current_size,
                    'total_neurons': self.current_size**2
                })
            else:
                print(f"Iteration {iteration + 1}:")
                print(f"  ‚è∏Ô∏è  No growth (error below threshold or max size reached)")
                print(f"  üìä Current grid: {self.current_size}√ó{self.current_size} = {self.current_size**2} neurons")
            print()

        print(f"\n‚úÖ GSOM growth simulation complete!")
        print(f"   Final grid size: {self.current_size}√ó{self.current_size} = {self.current_size**2} neurons")
        print(f"   Growth events: {len(self.growth_history)}")

        return self.growth_history

# Create and simulate Growing SOM
gsom = SimpleGrowingSOM(initial_size=3, max_size=10, growth_threshold=0.3, input_len=6)
growth_history = gsom.simulate_growth(X_normalized, num_iterations=5)

# Visualize growth history
if growth_history:
    print("\n" + "="*80)
    print("üìà GSOM GROWTH VISUALIZATION")
    print("="*80)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    # Plot 1: Grid size over iterations
    iterations = [h['iteration'] for h in growth_history]
    grid_sizes = [h['grid_size'] for h in growth_history]

    ax1.plot(iterations, grid_sizes, 'o-', linewidth=2, markersize=10, color='steelblue')
    ax1.set_xlabel('Iteration', fontsize=12, fontweight='bold')
    ax1.set_ylabel('Grid Size (n√ón)', fontsize=12, fontweight='bold')
    ax1.set_title('GSOM Grid Size Growth', fontsize=14, fontweight='bold')
    ax1.grid(True, alpha=0.3)
    ax1.set_ylim([2, gsom.max_size + 1])

    # Plot 2: Total neurons over iterations
    total_neurons = [h['total_neurons'] for h in growth_history]

    ax2.bar(iterations, total_neurons, color='coral', alpha=0.7, edgecolor='black')
    ax2.set_xlabel('Iteration', fontsize=12, fontweight='bold')
    ax2.set_ylabel('Total Neurons', fontsize=12, fontweight='bold')
    ax2.set_title('GSOM Total Neurons Growth', fontsize=14, fontweight='bold')
    ax2.grid(True, alpha=0.3, axis='y')

    plt.tight_layout()
    plt.show()

print("\nüí° GSOM Key Advantages:")
print("   ‚úì Automatic grid size determination")
print("   ‚úì Adapts to data complexity")
print("   ‚úì More efficient for sparse or non-uniform data")
print("   ‚úì Better representation of data distribution")
print("\n‚ö†Ô∏è  Note: This is a simplified simulation for educational purposes.")
print("   Real GSOM implementation requires error calculation and strategic neuron placement.")
````

### 11.2 Hierarchical Self-Organizing Map

**Concept**: Hierarchical SOM uses multiple SOM layers, where each layer provides progressively finer clustering.

**Advantages**:
- Multi-resolution clustering (coarse to fine)
- Better for large datasets
- Reveals hierarchical structure in data
- More interpretable for complex patterns

**How it works**:
1. **Level 1**: Train coarse SOM (e.g., 5√ó5) on all data
2. **Level 2**: For each Level 1 neuron, train fine SOM (e.g., 3√ó3) on data mapped to that neuron
3. Result: Hierarchical structure with coarse clusters subdivided into fine clusters

````python
# Hierarchical SOM Implementation (Two-Level)
print("="*80)
print("üèóÔ∏è  HIERARCHICAL SELF-ORGANIZING MAP (2-LEVEL)")
print("="*80)

# Level 1: Coarse SOM (5√ó5)
print("\nüìä LEVEL 1: Coarse SOM (5√ó5)")
print("-" * 80)

som_level1_size = 5  # Coarse grid
som_level1 = MiniSom(x=som_level1_size, y=som_level1_size,
                     input_len=input_len, sigma=2.0,
                     learning_rate=0.5, random_seed=42)

# Train Level 1 SOM
som_level1.random_weights_init(X_normalized)
som_level1.train_random(data=X_normalized, num_iteration=5000)

print(f"‚úÖ Level 1 SOM trained")
print(f"   Grid size: {som_level1_size}√ó{som_level1_size} = {som_level1_size**2} neurons")
print(f"   Quantization error: {som_level1.quantization_error(X_normalized):.4f}")

# Assign each data point to Level 1 neuron
level1_assignments = np.array([som_level1.winner(x) for x in X_normalized])

# Count data points per Level 1 neuron
level1_counts = {}
for winner in level1_assignments:
    key = tuple(winner)
    level1_counts[key] = level1_counts.get(key, 0) + 1

print(f"\nüìä Level 1 Distribution:")
print(f"   Active neurons: {len(level1_counts)} / {som_level1_size**2}")
print(f"   Average data points per active neuron: {np.mean(list(level1_counts.values())):.2f}")

# Level 2: Fine SOMs (3√ó3 for each Level 1 neuron with sufficient data)
print("\n" + "="*80)
print("üìä LEVEL 2: Fine SOMs (3√ó3 for each Level 1 cluster)")
print("-" * 80)

som_level2_size = 3  # Fine grid
min_data_points = 5  # Minimum data points to create Level 2 SOM
level2_soms = {}  # Dictionary to store Level 2 SOMs

# Create Level 2 SOM for each Level 1 neuron with sufficient data
for level1_neuron, count in level1_counts.items():
    if count >= min_data_points:  # Only create if enough data
        # Get data points assigned to this Level 1 neuron
        mask = np.all(level1_assignments == level1_neuron, axis=1)
        level1_data = X_normalized[mask]

        # Create and train Level 2 SOM
        som_level2 = MiniSom(x=som_level2_size, y=som_level2_size,
                            input_len=input_len, sigma=1.0,
                            learning_rate=0.3, random_seed=42)
        som_level2.random_weights_init(level1_data)
        som_level2.train_random(data=level1_data, num_iteration=2000)

        # Store Level 2 SOM
        level2_soms[level1_neuron] = {
            'som': som_level2,
            'data': level1_data,
            'count': count
        }

        print(f"‚úÖ Level 2 SOM created for Level 1 neuron {level1_neuron}")
        print(f"   Data points: {count}")
        print(f"   Grid size: {som_level2_size}√ó{som_level2_size} = {som_level2_size**2} neurons")
        print(f"   Quantization error: {som_level2.quantization_error(level1_data):.4f}")
        print()

print(f"\n‚úÖ Hierarchical SOM complete!")
print(f"   Level 1: {som_level1_size}√ó{som_level1_size} = {som_level1_size**2} coarse clusters")
print(f"   Level 2: {len(level2_soms)} fine SOMs created")
print(f"   Total fine clusters: {len(level2_soms) * som_level2_size**2}")
print(f"   Hierarchical depth: 2 levels")
````

````python
# Visualize Hierarchical SOM Structure
print("\n" + "="*80)
print("üìä HIERARCHICAL SOM VISUALIZATION")
print("="*80)

fig = plt.figure(figsize=(18, 8))
gs = GridSpec(2, 3, figure=fig, hspace=0.3, wspace=0.3)

# Plot 1: Level 1 U-Matrix
ax1 = fig.add_subplot(gs[0, :])
distance_map_l1 = som_level1.distance_map()
im1 = ax1.pcolor(distance_map_l1.T, cmap='bone_r', alpha=0.8)
plt.colorbar(im1, ax=ax1, label='Distance to Neighbors')
ax1.set_title('Level 1: Coarse SOM (5√ó5) - U-Matrix', fontsize=14, fontweight='bold')
ax1.set_xlabel('SOM Grid X', fontsize=11)
ax1.set_ylabel('SOM Grid Y', fontsize=11)

# Add hit counts to Level 1
for (x, y), count in level1_counts.items():
    ax1.text(x + 0.5, y + 0.5, str(count),
            ha='center', va='center', fontsize=12, fontweight='bold',
            color='white', bbox=dict(boxstyle='round', facecolor='red', alpha=0.7))

# Plot 2-4: Sample Level 2 SOMs (first 3)
for idx, (level1_neuron, level2_info) in enumerate(list(level2_soms.items())[:3]):
    ax = fig.add_subplot(gs[1, idx])

    # Get Level 2 U-Matrix
    distance_map_l2 = level2_info['som'].distance_map()
    im = ax.pcolor(distance_map_l2.T, cmap='viridis', alpha=0.8)
    plt.colorbar(im, ax=ax, label='Distance')

    ax.set_title(f'Level 2 SOM for L1 Neuron {level1_neuron}\n({level2_info["count"]} data points)',
                fontsize=11, fontweight='bold')
    ax.set_xlabel('Grid X', fontsize=10)
    ax.set_ylabel('Grid Y', fontsize=10)

plt.suptitle('Hierarchical SOM: Two-Level Structure', fontsize=16, fontweight='bold', y=0.98)
plt.show()

print("\n‚úÖ Hierarchical SOM visualization complete!")
print("\nüí° Hierarchical SOM Interpretation:")
print("   üìä Level 1 (Top): Coarse clustering into major groups")
print("   üìä Level 2 (Bottom): Fine clustering within each major group")
print("   üéØ Advantage: Multi-resolution view of data structure")
print("   üéØ Use case: Large datasets with hierarchical patterns")
````
