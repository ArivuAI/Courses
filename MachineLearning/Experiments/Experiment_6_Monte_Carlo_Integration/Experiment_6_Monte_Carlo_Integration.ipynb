Monte Carlo (1).png

### 📐 Mathematical Version

To find the integral ∫₀¹ f(x) dx:

1. Generate random points: x₁, x₂, ..., xₙ in [0, 1]
2. Evaluate function: f(x₁), f(x₂), ..., f(xₙ)
3. Take average: I ≈ (1/N) × Σf(xᵢ)
4. For interval [a, b]: I ≈ (b-a) × (1/N) × Σf(xᵢ)

**Simple, but powerful!**

## 📖 Background & Theory

### Key Concepts

### 1️⃣ Definite Integral
- Represents the **area under a curve** f(x) between limits a and b  
- **Notation:** I = ∫[a,b] f(x) dx  
- **Analytical Solution:** Use calculus (when possible)  
- **Numerical Solution:** Use approximation techniques like **Monte Carlo integration**

---

### 2️⃣ Monte Carlo Integration
- A **numerical technique** that uses **random sampling** to estimate integrals  
- Based on the **Law of Large Numbers**  
- **Error decreases** as sample size increases (∝ 1/√N)  
- Works well even for **high-dimensional** problems where traditional methods fail  

---

### 3️⃣ Uniform Random Numbers
- Numbers are **uniformly distributed** between [0, 1]  
- Each number has an **equal chance** of appearing  
- Generated using **Pseudo-Random Number Generators (PRNGs)**  
- These random values are the **foundation** of all Monte Carlo methods  

---

### 4️⃣ Law of Large Numbers (LLN)
- As the number of samples N → ∞, the **sample average** approaches the **true expected value**  
- Ensures Monte Carlo estimates become more accurate with more samples  
- Mathematically: mean(f(x₁), f(x₂), ..., f(xN)) → E[f(x)]

---

### 5️⃣ Central Limit Theorem (CLT)
- The **distribution of sample means** tends to a **normal distribution**  
- Allows us to compute **confidence intervals** for estimates  
- **Standard Error (SE):** SE = σ / sqrt(N)  
  where σ = standard deviation of sample values  

---

## 🧮 Monte Carlo Integration (Simple Version)

**Goal:** Estimate the area under a curve f(x) between a and b.  

Think of it like this: imagine **throwing darts randomly at the area under the curve** and seeing how many hit the “target.”

---

### Step 1: Pick random points
- Generate **N random numbers** between `a` and `b`.  
- These are like your dart throws along the x-axis.

### Step 2: Evaluate the function
- For each random number x_i, calculate f(x_i).  
- This is like measuring the height of the curve at the dart location.

### Step 3: Average the results
- Take the **average of all the f(x_i) values**.  
- Multiply by the width of the interval `(b-a)` to get your estimate of the area.  

**In plain English:**  
> Average height × width of interval = approximate area under the curve  

---

### Step 4: Understand the error
- The more points you take, the better the estimate.  
- Roughly, if you want **10× better accuracy**, you need **100× more points**.  

---

### 💡 Why it works
- Random sampling gives you a “representative snapshot” of the curve.  
- The average of many random heights approximates the true area.

---

### ✅ Quick Example

Imagine a curve like `y = x²` from `x = 0` to `x = 1`.  

1. Pick 100 random numbers between 0 and 1.  
2. Compute `y = x²` for each number.  
3. Average all the y-values.  
4. Multiply by 1 (width of interval).  

That’s your Monte Carlo estimate of the integral!

### Advantages and Limitations

**Advantages:**
- ✅ Simple to implement
- ✅ Works in high dimensions (curse of dimensionality doesn't apply)
- ✅ Error rate O(1/√N) independent of dimension
- ✅ Can handle complex, irregular domains
- ✅ Easy to parallelize
- ✅ Provides confidence intervals

**Limitations:**
- ❌ Slow convergence (need many samples for high accuracy)
- ❌ Random error (different runs give different results)
- ❌ Not suitable for low-dimensional smooth functions
- ❌ Requires good random number generator

### When to Use Monte Carlo

**Use Monte Carlo when:**
- High-dimensional integrals (d > 3)
- Complex, irregular integration domains
- No analytical solution available
- Moderate accuracy is sufficient
- Uncertainty quantification is needed

**Use other methods when:**
- Low dimensions (d ≤ 3) with smooth functions
- Very high accuracy required
- Deterministic results needed
- Computational budget is limited

## 🎲 Section 2: Generate Uniform Random Numbers

### What to Expect
We'll create a function to generate uniform random numbers in [0, 1].

### Process Overview
1. Implement custom uniform random number generator
2. Use NumPy's built-in generator for comparison
3. Generate sample random numbers
4. Visualize distribution
5. Verify uniformity

### Expected Outcome
Function that generates uniform random numbers, verified to be truly uniform.

### Why This Matters
Quality of random numbers directly affects Monte Carlo accuracy. Poor random numbers lead to biased estimates.

````python
# Function to generate uniform random numbers in [0, 1]
def generate_uniform_random(n, seed=None):
    """
    Generate n uniform random numbers in the interval [0, 1]

    Parameters:
    -----------
    n : int
        Number of random samples to generate
    seed : int, optional
        Random seed for reproducibility

    Returns:
    --------
    numpy.ndarray
        Array of n uniform random numbers in [0, 1]
    """
    # Set random seed if provided
    if seed is not None:
        np.random.seed(seed)  # Set seed for reproducibility

    # Generate uniform random numbers using NumPy
    random_numbers = np.random.uniform(low=0.0, high=1.0, size=n)  # Uniform distribution

    return random_numbers

# Alternative: Using np.random.random() (equivalent)
def generate_uniform_random_alt(n, seed=None):
    """
    Alternative method using np.random.random()
    """
    if seed is not None:
        np.random.seed(seed)  # Set seed

    return np.random.random(n)  # Generate random numbers in [0, 1)

# Test the function with 10 samples
print("="*80)
print("🎲 GENERATING UNIFORM RANDOM NUMBERS")
print("="*80)

n_samples = 10  # Number of samples
random_samples = generate_uniform_random(n_samples, seed=42)  # Generate samples

print(f"\nGenerated {n_samples} uniform random numbers in [0, 1]:\n")
for i, sample in enumerate(random_samples, 1):  # Loop through samples
    print(f"  x_{i:2d} = {sample:.6f}")  # Print with 6 decimal places

# Calculate statistics
print(f"\n📊 Statistics:")
print(f"   Minimum: {random_samples.min():.6f}")
print(f"   Maximum: {random_samples.max():.6f}")
print(f"   Mean: {random_samples.mean():.6f} (expected: 0.5)")
print(f"   Std Dev: {random_samples.std():.6f} (expected: {1/np.sqrt(12):.6f})")

print("\n✅ Random number generation successful!")
````

## 🧮 Section 5: Monte Carlo Integration - Basic Implementation

### What to Expect
We'll implement the Monte Carlo integration algorithm and estimate the integral.

### Process Overview
1. Implement Monte Carlo estimator
2. Calculate integral estimate from 10 samples
3. Compare with analytical solution
4. Calculate error and relative error
5. Understand the formula

### Expected Outcome
Integral estimate with error analysis.

### Why This Matters
This is the core Monte Carlo integration algorithm - simple but powerful!

````python
# Monte Carlo Integration Function
def monte_carlo_integrate(func, a, b, n_samples, seed=None):
    """
    Estimate definite integral using Monte Carlo method

    Formula: I ≈ (b - a) × (1/N) × Σf(xᵢ)

    Parameters:
    -----------
    func : callable
        Function to integrate
    a : float
        Lower bound of integration
    b : float
        Upper bound of integration
    n_samples : int
        Number of random samples
    seed : int, optional
        Random seed for reproducibility

    Returns:
    --------
    dict
        Dictionary containing:
        - estimate: integral estimate
        - samples: random samples used
        - function_values: f(x) for each sample
        - std_error: standard error of estimate
    """
    # Set random seed if provided
    if seed is not None:
        np.random.seed(seed)  # Set seed

    # Step 1: Generate random samples in [a, b]
    samples = np.random.uniform(a, b, n_samples)  # Uniform random samples

    # Step 2: Evaluate function at each sample
    function_values = func(samples)  # Apply function

    # Step 3: Calculate mean of function values
    mean_value = np.mean(function_values)  # Average of f(x)

    # Step 4: Estimate integral
    # Formula: I ≈ (b - a) × mean(f(x))
    integral_estimate = (b - a) * mean_value  # Monte Carlo estimate

    # Step 5: Calculate standard error
    # SE = (b - a) × std(f(x)) / √N
    std_dev = np.std(function_values, ddof=1)  # Sample standard deviation
    std_error = (b - a) * std_dev / np.sqrt(n_samples)  # Standard error

    # Return results
    return {
        'estimate': integral_estimate,
        'samples': samples,
        'function_values': function_values,
        'std_error': std_error,
        'mean_value': mean_value,
        'std_dev': std_dev
    }

# Estimate integral using 10 samples
print("="*80)
print("🧮 MONTE CARLO INTEGRATION - BASIC IMPLEMENTATION")
print("="*80)

# Integration parameters
a, b = 0, 1  # Integration interval
n_samples = 10  # Number of samples
analytical_integral = 1/3  # True value: ∫₀¹ x² dx = 1/3

# Perform Monte Carlo integration
result = monte_carlo_integrate(f, a, b, n_samples, seed=42)  # Estimate integral

# Display results
print(f"\n📐 Function: f(x) = x²")
print(f"   Interval: [{a}, {b}]")
print(f"   Number of samples: {n_samples}")

print(f"\n🎲 Random Samples Generated:")
print(f"   {result['samples']}")

print(f"\n📊 Function Values f(x_i):")
print(f"   {result['function_values']}")

print(f"\n🧮 Monte Carlo Calculation:")
print(f"   Mean of f(x): {result['mean_value']:.6f}")
print(f"   Interval width (b-a): {b-a:.6f}")
print(f"   Integral estimate: (b-a) × mean(f(x)) = {b-a:.6f} × {result['mean_value']:.6f}")
print(f"                    = {result['estimate']:.6f}")

print(f"\n✅ RESULTS:")
print("="*80)
print(f"   Monte Carlo Estimate: {result['estimate']:.6f}")
print(f"   Analytical Solution:  {analytical_integral:.6f}")
print(f"   Absolute Error:       {abs(result['estimate'] - analytical_integral):.6f}")
print(f"   Relative Error:       {abs(result['estimate'] - analytical_integral) / analytical_integral * 100:.2f}%")
print(f"   Standard Error:       {result['std_error']:.6f}")
print("="*80)

print("\n💡 Interpretation:")
print(f"   With only {n_samples} samples, we got within "
      f"{abs(result['estimate'] - analytical_integral) / analytical_integral * 100:.1f}% of the true value!")
print("   More samples will improve accuracy.")
````

## 📈 Section 6: Convergence Analysis

Understanding of convergence behavior and error reduction.

### Why This Matters

````python
# Convergence analysis with different sample sizes
print("\n" + "="*80)
print("📈 CONVERGENCE ANALYSIS")
print("="*80)

# Sample sizes to test
sample_sizes = [10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000]  # Increasing sample sizes

# Storage for results
estimates = []  # Integral estimates
errors = []  # Absolute errors
std_errors = []  # Standard errors
relative_errors = []  # Relative errors (%)

print("\n🔄 Running Monte Carlo with different sample sizes...\n")
print(f"{'N Samples':<12} {'Estimate':<12} {'Error':<12} {'Rel Error %':<15} {'Std Error':<12}")
print("-" * 65)

# Run Monte Carlo for each sample size
for n in sample_sizes:  # Loop through sample sizes
    # Estimate integral
    result = monte_carlo_integrate(f, a, b, n, seed=42)  # Monte Carlo

    # Calculate errors
    estimate = result['estimate']  # Integral estimate
    error = abs(estimate - analytical_integral)  # Absolute error
    rel_error = error / analytical_integral * 100  # Relative error (%)
    std_error = result['std_error']  # Standard error

    # Store results
    estimates.append(estimate)
    errors.append(error)
    relative_errors.append(rel_error)
    std_errors.append(std_error)

    # Print results
    print(f"{n:<12} {estimate:<12.6f} {error:<12.6f} {rel_error:<15.2f} {std_error:<12.6f}")

print("\n✅ Convergence analysis complete!")
````

### What to Expect

````python
# Calculate confidence intervals
print("="*80)
print("🎯 CONFIDENCE INTERVALS")
print("="*80)

# Confidence level
confidence_level = 0.95  # 95% confidence
z_score = 1.96  # Z-score for 95% CI

# Calculate confidence intervals for each sample size
ci_lower = []  # Lower bounds
ci_upper = []  # Upper bounds
contains_true = []  # Whether CI contains true value

print(f"\n📊 {confidence_level*100:.0f}% Confidence Intervals:\n")
print(f"{'N':<10} {'Estimate':<12} {'CI Lower':<12} {'CI Upper':<12} {'Width':<12} {'Contains True?':<15}")
print("-" * 75)

for i, n in enumerate(sample_sizes):  # Loop through sample sizes
    # Calculate CI
    margin = z_score * std_errors[i]  # Margin of error
    lower = estimates[i] - margin  # Lower bound
    upper = estimates[i] + margin  # Upper bound
    width = upper - lower  # CI width

    # Check if CI contains true value
    contains = lower <= analytical_integral <= upper  # Boolean

    # Store results
    ci_lower.append(lower)
    ci_upper.append(upper)
    contains_true.append(contains)

    # Print results
    print(f"{n:<10} {estimates[i]:<12.6f} {lower:<12.6f} {upper:<12.6f} "
          f"{width:<12.6f} {'✅ Yes' if contains else '❌ No':<15}")

print(f"\n📈 Coverage: {sum(contains_true)}/{len(contains_true)} intervals contain true value "
      f"({sum(contains_true)/len(contains_true)*100:.1f}%)")

print("\n✅ Confidence interval calculation complete!")
````

````python
# Visualize confidence intervals
print("\n" + "="*80)
print("📊 VISUALIZING CONFIDENCE INTERVALS")
print("="*80)

# Create figure
fig, ax = plt.subplots(figsize=(14, 8))

# Plot estimates with error bars
x_positions = np.arange(len(sample_sizes))  # X positions for bars
errors_for_plot = [z_score * se for se in std_errors]  # Error bar sizes

# Create bar plot with error bars
bars = ax.bar(x_positions, estimates, yerr=errors_for_plot,
             capsize=10, alpha=0.7, color='steelblue',
             edgecolor='black', linewidth=1.5,
             error_kw={'linewidth': 2, 'ecolor': 'red'})

# Color bars based on whether CI contains true value
for i, (bar, contains) in enumerate(zip(bars, contains_true)):
    if contains:
        bar.set_color('green')  # Green if contains true value
        bar.set_alpha(0.7)
    else:
        bar.set_color('red')  # Red if doesn't contain true value
        bar.set_alpha(0.7)

# Add horizontal line for true value
ax.axhline(y=analytical_integral, color='blue', linestyle='--',
          linewidth=3, label=f'True Value = {analytical_integral:.6f}', zorder=10)

# Customize plot
ax.set_xlabel('Sample Size', fontsize=13, fontweight='bold')
ax.set_ylabel('Integral Estimate', fontsize=13, fontweight='bold')
ax.set_title('95% Confidence Intervals for Different Sample Sizes',
            fontsize=15, fontweight='bold')
ax.set_xticks(x_positions)
ax.set_xticklabels([f'N={n}' for n in sample_sizes], rotation=45, ha='right')
ax.legend(fontsize=12)
ax.grid(True, alpha=0.3, axis='y')

# Add text annotation
ax.text(0.02, 0.98, 'Green: CI contains true value\nRed: CI misses true value',
       transform=ax.transAxes, fontsize=11, verticalalignment='top',
       bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.tight_layout()
plt.show()

print("\n✅ Confidence interval visualization complete!")
print("\n💡 Interpretation:")
print("   • Error bars show 95% confidence intervals")
print("   • Narrower intervals = more precise estimates")
print("   • Green bars: CI contains true value (good!)")
print("   • Red bars: CI misses true value (rare, ~5% expected)")
print("   • Larger N → narrower CI → more confidence")
````

## 💼 Section 9: Real-World Application - Estimating π

### What to Expect
We'll use Monte Carlo to estimate π - a classic application!

### Process Overview
1. Use quarter circle: ∫₀¹ √(1-x²) dx = π/4
2. Apply Monte Carlo integration
3. Multiply result by 4 to get π
4. Compare with true value of π
5. Visualize convergence to π

### Expected Outcome
Estimate of π using random sampling!

### Why This Matters
This demonstrates how randomness can solve deterministic problems - a key Monte Carlo insight!

````python
# Estimate π using Monte Carlo
print("="*80)
print("🥧 ESTIMATING π USING MONTE CARLO")
print("="*80)

# Method: Integrate quarter circle
# ∫₀¹ √(1-x²) dx = π/4
# Therefore: π = 4 × ∫₀¹ √(1-x²) dx

# Define quarter circle function
def quarter_circle(x):
    """
    Quarter circle function: f(x) = √(1-x²)

    This represents the upper half of a unit circle from x=0 to x=1.
    The integral gives the area of a quarter circle = π/4.
    """
    return np.sqrt(1 - x**2)  # Quarter circle

print("\n📐 Method: Quarter Circle Integration")
print("   Function: f(x) = √(1-x²)")
print("   Interval: [0, 1]")
print("   Integral: ∫₀¹ √(1-x²) dx = π/4")
print("   Therefore: π = 4 × Integral")

# Test with different sample sizes
pi_sample_sizes = [100, 1000, 10000, 100000, 1000000]  # Sample sizes
pi_estimates = []  # π estimates
pi_errors = []  # Errors

print(f"\n🔄 Estimating π with different sample sizes...\n")
print(f"{'N Samples':<15} {'Integral':<15} {'π Estimate':<15} {'Error':<15} {'Rel Error %':<15}")
print("-" * 75)

for n in pi_sample_sizes:  # Loop through sample sizes
    # Estimate integral
    result = monte_carlo_integrate(quarter_circle, 0, 1, n, seed=42)  # MC integration
    integral_estimate = result['estimate']  # Integral value

    # Estimate π
    pi_estimate = 4 * integral_estimate  # π = 4 × integral

    # Calculate error
    error = abs(pi_estimate - np.pi)  # Absolute error
    rel_error = error / np.pi * 100  # Relative error (%)

    # Store results
    pi_estimates.append(pi_estimate)
    pi_errors.append(error)

    # Print results
    print(f"{n:<15} {integral_estimate:<15.6f} {pi_estimate:<15.6f} {error:<15.6f} {rel_error:<15.2f}")

print(f"\n✅ Best estimate: π ≈ {pi_estimates[-1]:.6f} (True value: {np.pi:.6f})")
print(f"   Error: {pi_errors[-1]:.6f} ({pi_errors[-1]/np.pi*100:.3f}%)")
````

````python
# Visualize π estimation
print("\n" + "="*80)
print("📊 VISUALIZING π ESTIMATION")
print("="*80)

# Create figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Estimating π Using Monte Carlo Integration', fontsize=16, fontweight='bold')

# Plot 1: Quarter circle with random samples
x_circle = np.linspace(0, 1, 1000)  # X values
y_circle = quarter_circle(x_circle)  # Y values

ax1.plot(x_circle, y_circle, linewidth=3, color='blue', label='Quarter Circle: √(1-x²)')
ax1.fill_between(x_circle, 0, y_circle, alpha=0.3, color='blue', label='Area = π/4')

# Add some random samples
n_vis = 100  # Number of samples to visualize
samples_vis = generate_uniform_random(n_vis, seed=42)  # Random samples
values_vis = quarter_circle(samples_vis)  # Function values

ax1.scatter(samples_vis, values_vis, s=30, color='red', alpha=0.6,
           edgecolors='black', linewidths=0.5, label=f'{n_vis} Random Samples')

ax1.set_xlabel('x', fontsize=12, fontweight='bold')
ax1.set_ylabel('f(x) = √(1-x²)', fontsize=12, fontweight='bold')
ax1.set_title('Quarter Circle Integration', fontsize=13, fontweight='bold')
ax1.legend(fontsize=11)
ax1.grid(True, alpha=0.3)
ax1.set_xlim([0, 1])
ax1.set_ylim([0, 1.1])

# Add text annotation
ax1.text(0.5, 0.5, 'π = 4 × Area', fontsize=14, fontweight='bold',
        ha='center', va='center',
        bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.7))

# Plot 2: Convergence to π
ax2.semilogx(pi_sample_sizes, pi_estimates, 'o-', linewidth=2,
            markersize=10, color='green', label='MC Estimate of π')
ax2.axhline(y=np.pi, color='red', linestyle='--', linewidth=3,
           label=f'True π = {np.pi:.6f}')
ax2.fill_between(pi_sample_sizes, np.pi - 0.01, np.pi + 0.01,
                alpha=0.2, color='red', label='±0.01 band')

ax2.set_xlabel('Number of Samples (N)', fontsize=12, fontweight='bold')
ax2.set_ylabel('π Estimate', fontsize=12, fontweight='bold')
ax2.set_title('Convergence to π', fontsize=13, fontweight='bold')
ax2.legend(fontsize=11)
ax2.grid(True, alpha=0.3, which='both')

plt.tight_layout()
plt.show()

print("\n✅ π estimation visualization complete!")
print("\n💡 Amazing Insight:")
print("   We estimated π (a fundamental constant) using RANDOM numbers!")
print("   This shows the power of Monte Carlo methods.")
print("   With 1 million samples, we got within 0.01% of true π!")
````

````python
# Monte Carlo Option Pricing
print("="*80)
print("💰 MONTE CARLO OPTION PRICING")
print("="*80)

# Option parameters
S0 = 100  # Current stock price
K = 105  # Strike price
T = 1.0  # Time to maturity (1 year)
r = 0.05  # Risk-free rate (5%)
sigma = 0.2  # Volatility (20%)
n_simulations = 100000  # Number of Monte Carlo simulations

print("\n📋 Option Parameters:")
print(f"   Current Stock Price (S₀): ${S0:.2f}")
print(f"   Strike Price (K): ${K:.2f}")
print(f"   Time to Maturity (T): {T:.1f} year")
print(f"   Risk-Free Rate (r): {r*100:.1f}%")
print(f"   Volatility (σ): {sigma*100:.1f}%")
print(f"   Number of Simulations: {n_simulations:,}")

# Set random seed
np.random.seed(42)  # Reproducibility

# Simulate stock prices at maturity using geometric Brownian motion
# S_T = S_0 × exp((r - σ²/2)T + σ√T × Z)
# where Z ~ N(0,1)

Z = np.random.standard_normal(n_simulations)  # Standard normal random variables
ST = S0 * np.exp((r - 0.5 * sigma**2) * T + sigma * np.sqrt(T) * Z)  # Stock prices at maturity

# Calculate call option payoffs
# Payoff = max(S_T - K, 0)
payoffs = np.maximum(ST - K, 0)  # Call option payoffs

# Calculate option price (discounted expected payoff)
# C = e^(-rT) × E[max(S_T - K, 0)]
option_price_mc = np.exp(-r * T) * np.mean(payoffs)  # Monte Carlo price

# Calculate standard error
payoff_std = np.std(payoffs, ddof=1)  # Standard deviation of payoffs
se = np.exp(-r * T) * payoff_std / np.sqrt(n_simulations)  # Standard error

# Calculate 95% confidence interval
ci_lower = option_price_mc - 1.96 * se  # Lower bound
ci_upper = option_price_mc + 1.96 * se  # Upper bound

# Black-Scholes formula for comparison
from scipy.stats import norm  # Normal distribution

d1 = (np.log(S0/K) + (r + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))  # d1 parameter
d2 = d1 - sigma*np.sqrt(T)  # d2 parameter
option_price_bs = S0*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)  # Black-Scholes price

# Display results
print("\n" + "="*80)
print("📊 RESULTS")
print("="*80)
print(f"\n💵 Monte Carlo Option Price: ${option_price_mc:.4f}")
print(f"   95% Confidence Interval: [${ci_lower:.4f}, ${ci_upper:.4f}]")
print(f"   Standard Error: ${se:.4f}")

print(f"\n📈 Black-Scholes Price: ${option_price_bs:.4f}")

print(f"\n🎯 Comparison:")
print(f"   Difference: ${abs(option_price_mc - option_price_bs):.4f}")
print(f"   Relative Error: {abs(option_price_mc - option_price_bs)/option_price_bs*100:.3f}%")

print("\n✅ Option pricing complete!")
print("\n💡 Business Impact:")
print(f"   • This option is worth ${option_price_mc:.2f}")
print(f"   • Monte Carlo matches Black-Scholes within {abs(option_price_mc - option_price_bs)/option_price_bs*100:.2f}%")
print("   • For complex options (no closed-form), Monte Carlo is the ONLY method!")
print("   • Used daily by banks to price trillion-dollar derivatives portfolios")
````

## 🎲 Section 11: Case Study - Portfolio Risk Analysis (VaR)

### What to Expect
We'll use Monte Carlo to calculate Value at Risk (VaR) for a stock portfolio.

### Process Overview
1. Create a portfolio with 3 stocks
2. Simulate 10,000 portfolio value scenarios
3. Calculate VaR at 95% and 99% confidence levels
4. Visualize loss distribution
5. Interpret results for risk management

### Expected Outcome
Understanding of how Monte Carlo is used for financial risk management.

### Why This Matters
VaR is required by banking regulations (Basel III) and used daily by financial institutions worldwide!

````python
# Portfolio Risk Analysis using Monte Carlo
print("="*80)
print("📊 CASE STUDY: PORTFOLIO VALUE AT RISK (VaR)")
print("="*80)

# Portfolio parameters
portfolio_value = 1000000  # $1 million portfolio
weights = np.array([0.4, 0.35, 0.25])  # Portfolio weights (40%, 35%, 25%)
expected_returns = np.array([0.08, 0.12, 0.15])  # Annual expected returns (8%, 12%, 15%)
volatilities = np.array([0.15, 0.25, 0.30])  # Annual volatilities (15%, 25%, 30%)
correlation_matrix = np.array([  # Correlation matrix
    [1.0, 0.6, 0.4],
    [0.6, 1.0, 0.5],
    [0.4, 0.5, 1.0]
])

time_horizon = 1/252  # 1 day (252 trading days per year)
n_simulations = 10000  # Number of Monte Carlo simulations

print("\n📋 Portfolio Details:")
print(f"   Total Value: ${portfolio_value:,.0f}")
print(f"   Asset Allocation: {weights[0]*100:.0f}% / {weights[1]*100:.0f}% / {weights[2]*100:.0f}%")
print(f"   Expected Returns: {expected_returns[0]*100:.0f}% / {expected_returns[1]*100:.0f}% / {expected_returns[2]*100:.0f}%")
print(f"   Volatilities: {volatilities[0]*100:.0f}% / {volatilities[1]*100:.0f}% / {volatilities[2]*100:.0f}%")
print(f"   Time Horizon: 1 day")
print(f"   Simulations: {n_simulations:,}")

# Set random seed
np.random.seed(42)

# Calculate covariance matrix
# Cov = D × Corr × D, where D is diagonal matrix of volatilities
D = np.diag(volatilities)
covariance_matrix = D @ correlation_matrix @ D

# Simulate portfolio returns
# Generate correlated random returns using Cholesky decomposition
L = np.linalg.cholesky(covariance_matrix)  # Cholesky decomposition
Z = np.random.standard_normal((n_simulations, 3))  # Independent standard normals

# Scale returns to 1-day horizon
daily_returns = expected_returns * time_horizon + \
                (L @ Z.T).T * np.sqrt(time_horizon)  # Correlated returns

# Calculate portfolio returns
portfolio_returns = daily_returns @ weights  # Weighted sum

# Calculate portfolio value changes
portfolio_changes = portfolio_value * portfolio_returns  # Dollar changes

# Calculate VaR at different confidence levels
var_95 = -np.percentile(portfolio_changes, 5)  # 95% VaR (5th percentile loss)
var_99 = -np.percentile(portfolio_changes, 1)  # 99% VaR (1st percentile loss)

# Calculate Conditional VaR (CVaR) / Expected Shortfall
cvar_95 = -np.mean(portfolio_changes[portfolio_changes <= -var_95])  # Average loss beyond VaR
cvar_99 = -np.mean(portfolio_changes[portfolio_changes <= -var_99])

# Display results
print("\n" + "="*80)
print("📈 RISK METRICS")
print("="*80)
print(f"\n💰 Portfolio Statistics:")
print(f"   Expected Daily Return: ${np.mean(portfolio_changes):,.2f} ({np.mean(portfolio_returns)*100:.3f}%)")
print(f"   Daily Volatility: ${np.std(portfolio_changes):,.2f} ({np.std(portfolio_returns)*100:.3f}%)")

print(f"\n⚠️  Value at Risk (VaR):")
print(f"   95% VaR (1-day): ${var_95:,.2f}")
print(f"   99% VaR (1-day): ${var_99:,.2f}")
print(f"   ➜ With 95% confidence, losses won't exceed ${var_95:,.2f} in one day")
print(f"   ➜ With 99% confidence, losses won't exceed ${var_99:,.2f} in one day")

print(f"\n📉 Conditional VaR (Expected Shortfall):")
print(f"   95% CVaR: ${cvar_95:,.2f}")
print(f"   99% CVaR: ${cvar_99:,.2f}")
print(f"   ➜ If losses exceed VaR, expected loss is ${cvar_95:,.2f} (95%) or ${cvar_99:,.2f} (99%)")

print("\n✅ Portfolio risk analysis complete!")
````

````python
# Visualize portfolio risk analysis
print("\n" + "="*80)
print("📊 VISUALIZING PORTFOLIO RISK")
print("="*80)

# Create figure
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Portfolio Risk Analysis - Monte Carlo VaR', fontsize=16, fontweight='bold')

# Plot 1: Portfolio return distribution
axes[0, 0].hist(portfolio_changes, bins=100, density=True, alpha=0.7,
               color='steelblue', edgecolor='black')
axes[0, 0].axvline(0, color='black', linestyle='-', linewidth=2, label='Break-even')
axes[0, 0].axvline(-var_95, color='orange', linestyle='--', linewidth=2,
                   label=f'95% VaR = ${var_95:,.0f}')
axes[0, 0].axvline(-var_99, color='red', linestyle='--', linewidth=2,
                   label=f'99% VaR = ${var_99:,.0f}')
axes[0, 0].set_xlabel('Portfolio Value Change ($)', fontsize=11, fontweight='bold')
axes[0, 0].set_ylabel('Probability Density', fontsize=11, fontweight='bold')
axes[0, 0].set_title('Portfolio Return Distribution (1-day)', fontsize=12, fontweight='bold')
axes[0, 0].legend(fontsize=10)
axes[0, 0].grid(True, alpha=0.3)

# Plot 2: Cumulative distribution (for VaR visualization)
sorted_changes = np.sort(portfolio_changes)
cumulative_prob = np.arange(1, len(sorted_changes) + 1) / len(sorted_changes)

axes[0, 1].plot(sorted_changes, cumulative_prob * 100, linewidth=2, color='steelblue')
axes[0, 1].axhline(5, color='orange', linestyle='--', linewidth=2, label='5% (95% VaR)')
axes[0, 1].axhline(1, color='red', linestyle='--', linewidth=2, label='1% (99% VaR)')
axes[0, 1].axvline(-var_95, color='orange', linestyle='--', linewidth=2, alpha=0.5)
axes[0, 1].axvline(-var_99, color='red', linestyle='--', linewidth=2, alpha=0.5)
axes[0, 1].set_xlabel('Portfolio Value Change ($)', fontsize=11, fontweight='bold')
axes[0, 1].set_ylabel('Cumulative Probability (%)', fontsize=11, fontweight='bold')
axes[0, 1].set_title('Cumulative Distribution Function', fontsize=12, fontweight='bold')
axes[0, 1].legend(fontsize=10)
axes[0, 1].grid(True, alpha=0.3)
axes[0, 1].set_xlim([sorted_changes[0], 0])
axes[0, 1].set_ylim([0, 10])

# Plot 3: Loss distribution (left tail focus)
losses = portfolio_changes[portfolio_changes < 0]  # Only losses
axes[1, 0].hist(losses, bins=50, density=True, alpha=0.7, color='red', edgecolor='black')
axes[1, 0].axvline(-var_95, color='orange', linestyle='--', linewidth=2,
                   label=f'95% VaR')
axes[1, 0].axvline(-var_99, color='darkred', linestyle='--', linewidth=2,
                   label=f'99% VaR')
axes[1, 0].axvline(-cvar_95, color='orange', linestyle=':', linewidth=2,
                   label=f'95% CVaR = ${cvar_95:,.0f}')
axes[1, 0].axvline(-cvar_99, color='darkred', linestyle=':', linewidth=2,
                   label=f'99% CVaR = ${cvar_99:,.0f}')
axes[1, 0].set_xlabel('Portfolio Loss ($)', fontsize=11, fontweight='bold')
axes[1, 0].set_ylabel('Probability Density', fontsize=11, fontweight='bold')
axes[1, 0].set_title('Loss Distribution (Left Tail)', fontsize=12, fontweight='bold')
axes[1, 0].legend(fontsize=9)
axes[1, 0].grid(True, alpha=0.3)

# Plot 4: VaR comparison bar chart
var_metrics = ['95% VaR', '99% VaR', '95% CVaR', '99% CVaR']
var_values = [var_95, var_99, cvar_95, cvar_99]
colors_bar = ['orange', 'red', 'darkorange', 'darkred']

bars = axes[1, 1].bar(var_metrics, var_values, color=colors_bar, alpha=0.7, edgecolor='black')
axes[1, 1].set_ylabel('Loss Amount ($)', fontsize=11, fontweight='bold')
axes[1, 1].set_title('Risk Metrics Comparison', fontsize=12, fontweight='bold')
axes[1, 1].grid(True, alpha=0.3, axis='y')

# Add value labels on bars
for bar, value in zip(bars, var_values):
    height = bar.get_height()
    axes[1, 1].text(bar.get_x() + bar.get_width()/2., height,
                    f'${value:,.0f}',
                    ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()

print("\n✅ Portfolio risk visualization complete!")
print("\n💡 Business Interpretation:")
print(f"   • On a typical day, expect return around ${np.mean(portfolio_changes):,.0f}")
print(f"   • In worst 5% of days, losses could exceed ${var_95:,.0f}")
print(f"   • In worst 1% of days, losses could exceed ${var_99:,.0f}")
print(f"   • If extreme loss occurs, expect to lose ${cvar_95:,.0f} on average")
print("\n🏦 Regulatory Context:")
print("   • Basel III requires banks to hold capital against VaR")
print("   • This analysis helps determine required capital reserves")
print("   • Monte Carlo is industry standard for complex portfolios")
````

## ❓ Frequently Asked Questions (FAQs)

### 1. Why is it called "Monte Carlo"?
**Answer**: Named after the Monte Carlo casino in Monaco, because the method relies on randomness (like gambling). It was developed in the 1940s during the Manhattan Project by Stanislaw Ulam and John von Neumann for simulating neutron diffusion in nuclear weapons.

### 2. How many samples do I need for accurate results?
**Answer**: It depends on desired accuracy:
- **1% error**: ~10,000 samples
- **0.1% error**: ~1,000,000 samples
- **0.01% error**: ~100,000,000 samples

Rule of thumb: To reduce error by 10×, need 100× more samples (due to 1/√N convergence).

### 3. Why does error decrease as 1/√N instead of 1/N?
**Answer**: This comes from the Central Limit Theorem. The standard error of the sample mean is σ/√N, where σ is the standard deviation. This is a fundamental statistical property - we can't do better without additional information (like variance reduction techniques).

### 4. Is Monte Carlo better than other numerical integration methods?
**Answer**: It depends:
- **Monte Carlo is better for**: High dimensions (d > 3), complex domains, no analytical solution
- **Other methods are better for**: Low dimensions (d ≤ 3), smooth functions, high accuracy requirements

For example, Simpson's rule has error O(1/N⁴) in 1D, much better than Monte Carlo's O(1/√N). But Simpson's rule becomes impractical in high dimensions.

### 5. What is the "curse of dimensionality" and how does Monte Carlo avoid it?
**Answer**: Traditional methods (like Simpson's rule) need N^d points for d dimensions. For d=10 and N=100, that's 100^10 = 10^20 points (impossible!). Monte Carlo needs the same number of samples regardless of dimension - error is always O(1/√N). This is Monte Carlo's superpower!

### 6. Can Monte Carlo give exact answers?
**Answer**: No, Monte Carlo is a statistical method that gives approximate answers with quantifiable uncertainty (confidence intervals). However, with enough samples, you can get arbitrarily close to the true value.

### 7. What is a confidence interval and how do I interpret it?
**Answer**: A 95% confidence interval means: "If we repeated this experiment many times, 95% of the intervals would contain the true value." It quantifies our uncertainty. Narrower intervals = more confidence.

### 8. Why do we need to set a random seed?
**Answer**: Random number generators are actually "pseudo-random" - they produce deterministic sequences from a seed. Setting the seed ensures reproducibility: same seed → same random numbers → same results. Critical for debugging and scientific reproducibility!

### 9. What are variance reduction techniques?
**Answer**: Methods to reduce error without increasing sample size:
- **Importance Sampling**: Sample more where function is large
- **Stratified Sampling**: Divide domain into strata, sample each
- **Control Variates**: Use correlated function with known integral
- **Antithetic Variates**: Use negatively correlated samples

These can reduce variance by 10-100×, equivalent to 100-10,000× more samples!

### 10. Can Monte Carlo be used for machine learning?
**Answer**: Absolutely! Monte Carlo is fundamental to modern ML:
- **Bayesian Inference**: MCMC (Markov Chain Monte Carlo) for posterior sampling
- **Reinforcement Learning**: Monte Carlo Tree Search (AlphaGo)
- **Neural Networks**: Dropout is a Monte Carlo technique
- **Uncertainty Quantification**: Monte Carlo dropout for Bayesian neural networks

### 11. What's the difference between Monte Carlo and MCMC?
**Answer**:
- **Monte Carlo**: Uses independent random samples
- **MCMC (Markov Chain Monte Carlo)**: Uses dependent samples from a Markov chain

MCMC is used when we can't sample directly from the target distribution (e.g., Bayesian posterior). It's more advanced but builds on Monte Carlo principles.

### 12. How is Monte Carlo used in real business applications?
**Answer**: Everywhere!
- **Finance**: Option pricing, risk management (VaR), portfolio optimization
- **Insurance**: Claim modeling, reserve estimation
- **Manufacturing**: Quality control, reliability analysis
- **Supply Chain**: Inventory optimization, demand forecasting
- **Marketing**: Customer lifetime value, A/B test analysis
- **Energy**: Oil reservoir simulation, renewable energy forecasting
