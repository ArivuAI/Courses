### üßÆ Mathematical Foundation of K-Means Clustering

---

## **1. The Core Objective: Within-Cluster Sum of Squares (WCSS)**

The primary goal of **K-Means** is to find the best possible set of **cluster centers** ($\mu$) and **assignments** ($S$) that minimize the total squared distance between every point and the center of the cluster to which that point is assigned.

This minimized value is known as **Inertia** or the **Within-Cluster Sum of Squares (WCSS)**.

$$
\text{Minimize: } WCSS = \sum_{j=1}^{K} \sum_{x \in S_j} \|x - \mu_j\|^2
$$

Where:  
- $K$: The predetermined number of clusters  
- $S_j$: The set of data points belonging to cluster $j$  
- $\mu_j$: The centroid (mean vector) of cluster $j$  
- $\|x - \mu_j\|^2$: The squared Euclidean distance between the data point $x$ and its assigned centroid $\mu_j$

**Intuitive Goal:**  
The smaller the WCSS, the more compact and cohesive the clusters are.

---

## **2. The Distance Metric: Euclidean Distance**

K-Means uses a **distance metric** to determine which cluster a point belongs to.  
The standard metric is the **Euclidean Distance**, which measures the straight-line distance between a data point and a cluster centroid in the feature space.

The Euclidean Distance between a point $x$ and a centroid $c$ in an $n$-dimensional space is:

$$
d(x, c) = \sqrt{\sum_{i=1}^{n} (x_i - c_i)^2}
$$

Where:  
- $x_i$ and $c_i$: Coordinate values of the data point and centroid for feature $i$  
- $n$: Total number of features (dimensions) in the dataset (e.g., $n=2$ for *Income* and *Score*)

**Key Takeaway:**  
By using the **squared distance** in the WCSS formula ($\|x - \mu_j\|^2$), the algorithm heavily penalizes larger distances, encouraging **tighter and well-separated clusters**.

---

## **3. The Role of WCSS in the Elbow Method**

The **Elbow Method** helps determine the optimal number of clusters ($K$).  
It involves plotting the WCSS (Inertia) for a range of $K$ values (e.g., $K = 1$ to $10$).

The "elbow" point on the curve indicates where adding another cluster yields **diminishing returns** ‚Äî  
i.e., minimal improvement in reducing WCSS ‚Äî thus balancing model simplicity and cluster compactness.

---


### **K-Means Algorithm Steps:**

1. Initialize: Randomly select K points as initial centroids
2. Assignment: Assign each point to nearest centroid
3. Update: Recalculate centroids as mean of assigned points
4. Repeat steps 2-3 until convergence (centroids don't change)

### Determining Optimal K

**1. Elbow Method**
- Plot within-cluster sum of squares (WCSS) vs K
- Look for "elbow" point where improvement slows
- Trade-off between number of clusters and compactness

**2. Silhouette Score**
- Measures how similar a point is to its own cluster vs other clusters
- Range: -1 to 1 (higher is better)
- Score > 0.5 indicates good clustering

**3. Business Constraints**
- Too few clusters: Lose personalization
- Too many clusters: Hard to manage campaigns
- Practical sweet spot: 3-7 segments for marketing

## üîß Section 4: Data Preprocessing and Feature Selection

```python
# Initialize StandardScaler for feature normalization
scaler = StandardScaler()  # Create scaler object

# Fit scaler to data and transform features
# StandardScaler: (x - mean) / std_dev ‚Üí transforms to mean=0, std=1
X_scaled = scaler.fit_transform(X)  # Fit and transform in one step

print("\n" + "="*80)
print("‚öôÔ∏è  FEATURE SCALING APPLIED")
print("="*80)
print("Scaling Method: StandardScaler (Z-score normalization)")
print("Formula: X_scaled = (X - mean) / standard_deviation")
print("\nResult: All features now have mean ‚âà 0 and standard deviation ‚âà 1")

# Display scaled feature statistics
print("\n" + "="*80)
print("üìä SCALED FEATURE RANGES (After Scaling)")
print("="*80)
for idx, feature in enumerate(features_for_clustering):  # Loop through each feature
    min_val = X_scaled[:, idx].min()  # Minimum value
    max_val = X_scaled[:, idx].max()  # Maximum value
    mean_val = X_scaled[:, idx].mean()  # Mean value (should be ~0)
    std_val = X_scaled[:, idx].std()  # Standard deviation (should be ~1)
    print(f"{feature:20s}: Min={min_val:7.2f}, Max={max_val:7.2f}, Mean={mean_val:7.2f}, Std={std_val:7.2f}")

print("\n‚úÖ All features now on the same scale!")
print("üí° Each feature will contribute equally to distance calculations")
print("üéØ Data is ready for K-Means clustering!")
```

## üîç Section 5: Determining Optimal Number of Clusters (K)

### What to Expect
We'll use two methods to find the optimal number of clusters: Elbow Method and Silhouette Analysis.

### Process Overview
1. **Elbow Method**: Test K from 2 to 10, plot WCSS (Within-Cluster Sum of Squares)
2. **Silhouette Analysis**: Calculate silhouette scores for different K values
3. Compare results from both methods
4. Choose optimal K based on both metrics and business needs

### Expected Outcome
- Elbow plot showing optimal K around 4-5
- Silhouette scores confirming the choice
- Clear decision on number of customer segments

### Why This Matters
Choosing the right K is crucial! Too few clusters = lose personalization. Too many clusters = hard to manage campaigns. We need the sweet spot!

```python
# Method 1: Elbow Method - Find optimal K by plotting WCSS
print("="*80)
print("üìà METHOD 1: ELBOW METHOD")
print("="*80)
print("Testing K values from 2 to 10...\n")

# Initialize lists to store results
k_values = range(2, 11)  # Test K from 2 to 10
wcss_values = []  # Within-Cluster Sum of Squares (inertia)

# Calculate WCSS for each K value
for k in k_values:  # Loop through each K value
    # Create K-Means model with k clusters
    kmeans = KMeans(n_clusters=k,  # Number of clusters
                   init='k-means++',  # Smart initialization (faster convergence).
                   n_init=10,  # Number of times to run with different initializations
                   max_iter=300,  # Maximum iterations per run
                   random_state=42)  # For reproducibility

    # Fit model to scaled data
    kmeans.fit(X_scaled)  # Train K-Means

    # Store WCSS (inertia) value
    wcss_values.append(kmeans.inertia_)  # Sum of squared distances to nearest cluster center

    print(f"K = {k:2d} ‚Üí WCSS = {kmeans.inertia_:8.2f}")

print("\n‚úÖ WCSS calculated for all K values!")
```

```python
# Plot Elbow Curve
plt.figure(figsize=(12, 6))  # Set figure size

# Create line plot
plt.plot(k_values, wcss_values,  # X and Y data
        marker='o',  # Circle markers
        markersize=10,  # Marker size
        linewidth=2,  # Line width
        color='steelblue',  # Line color
        markerfacecolor='orange',  # Marker fill color
        markeredgecolor='black',  # Marker edge color
        markeredgewidth=1.5)  # Marker edge width

# Add labels and title
plt.xlabel('Number of Clusters (K)', fontsize=14, fontweight='bold')  # X-axis label
plt.ylabel('WCSS (Within-Cluster Sum of Squares)', fontsize=14, fontweight='bold')  # Y-axis label
plt.title('Elbow Method: Finding Optimal K', fontsize=16, fontweight='bold')  # Title
plt.grid(True, alpha=0.3)  # Add grid
plt.xticks(k_values)  # Show all K values on x-axis

# Add annotation for elbow point
plt.annotate('Look for the "elbow" point\nwhere curve starts to flatten',  # Annotation text
            xy=(4, wcss_values[2]),  # Point to annotate (K=4)
            xytext=(6, wcss_values[0]),  # Text position
            arrowprops=dict(arrowstyle='->', color='red', lw=2),  # Arrow properties
            fontsize=12,  # Font size
            bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.7))  # Text box

plt.tight_layout()  # Adjust layout
plt.show()  # Display plot

print("\nüìä Elbow Plot Interpretation:")
print("   ‚Ä¢ The 'elbow' is where adding more clusters doesn't significantly reduce WCSS")
print("   ‚Ä¢ This suggests the optimal number of clusters")
print("   ‚Ä¢ From the plot, K = 4 or K = 5 appears to be the elbow point")
```

```python
# Method 2: Silhouette Analysis - Measure cluster quality
print("\n" + "="*80)
print("üìä METHOD 2: SILHOUETTE ANALYSIS")
print("="*80)
print("Calculating silhouette scores for K = 2 to 10...\n")

# Initialize list to store silhouette scores
silhouette_scores = []  # Silhouette score for each K

# Calculate silhouette score for each K value
for k in k_values:  # Loop through each K value
    # Create and fit K-Means model
    kmeans = KMeans(n_clusters=k,  # Number of clusters
                   init='k-means++',  # Smart initialization
                   n_init=10,  # Number of initializations
                   max_iter=300,  # Maximum iterations
                   random_state=42)  # For reproducibility

    # Get cluster labels
    cluster_labels = kmeans.fit_predict(X_scaled)  # Fit and predict in one step

    # Calculate silhouette score
    # Silhouette score measures how similar a point is to its own cluster vs other clusters
    # Range: -1 to 1 (higher is better)
    score = silhouette_score(X_scaled, cluster_labels)  # Calculate average silhouette score
    silhouette_scores.append(score)  # Store score

    print(f"K = {k:2d} ‚Üí Silhouette Score = {score:.4f}")

print("\n‚úÖ Silhouette scores calculated for all K values!")
```

```python
# Plot Silhouette Scores
plt.figure(figsize=(12, 6))  # Set figure size

# Create bar plot
bars = plt.bar(k_values, silhouette_scores,  # X and Y data
              color='lightcoral',  # Bar color
              edgecolor='black',  # Bar edge color
              linewidth=1.5,  # Edge width
              alpha=0.7)  # Transparency

# Highlight the best score
best_k_idx = silhouette_scores.index(max(silhouette_scores))  # Index of best score
best_k = k_values[best_k_idx]  # Best K value
bars[best_k_idx].set_color('green')  # Highlight best bar in green
bars[best_k_idx].set_alpha(0.9)  # Make it more opaque

# Add value labels on top of bars
for i, (k, score) in enumerate(zip(k_values, silhouette_scores)):  # Loop through bars
    plt.text(k, score + 0.01,  # Position (slightly above bar)
            f'{score:.3f}',  # Text (score value)
            ha='center',  # Horizontal alignment
            va='bottom',  # Vertical alignment
            fontsize=10,  # Font size
            fontweight='bold')  # Bold text

# Add labels and title
plt.xlabel('Number of Clusters (K)', fontsize=14, fontweight='bold')  # X-axis label
plt.ylabel('Silhouette Score', fontsize=14, fontweight='bold')  # Y-axis label
plt.title('Silhouette Analysis: Finding Optimal K', fontsize=16, fontweight='bold')  # Title
plt.grid(True, alpha=0.3, axis='y')  # Add horizontal grid
plt.xticks(k_values)  # Show all K values
plt.ylim(0, max(silhouette_scores) + 0.1)  # Set y-axis limits

# Add horizontal line for "good" threshold
plt.axhline(y=0.5, color='red', linestyle='--', linewidth=2, label='Good clustering threshold (0.5)')  # Reference line
plt.legend(fontsize=12)  # Add legend

plt.tight_layout()  # Adjust layout
plt.show()  # Display plot

print(f"\nüìä Silhouette Analysis Results:")
print(f"   ‚Ä¢ Best K = {best_k} with Silhouette Score = {max(silhouette_scores):.4f}")
print(f"   ‚Ä¢ Interpretation:")
print(f"     - Score > 0.7: Strong clustering")
print(f"     - Score > 0.5: Reasonable clustering")
print(f"     - Score > 0.25: Weak clustering")
print(f"     - Score < 0.25: No substantial clustering")
```

```python
# Decision: Choose optimal K based on both methods
print("\n" + "="*80)
print("üéØ OPTIMAL K DECISION")
print("="*80)
print("\nüìä Summary of Both Methods:")
print(f"   ‚Ä¢ Elbow Method suggests: K = 4 or K = 5")
print(f"   ‚Ä¢ Silhouette Analysis suggests: K = {best_k} (score = {max(silhouette_scores):.4f})")
print(f"\nüíº Business Perspective:")
print(f"   ‚Ä¢ K = 4 provides manageable number of customer segments")
print(f"   ‚Ä¢ Each segment can have dedicated marketing strategy")
print(f"   ‚Ä¢ Not too few (lose personalization) or too many (hard to manage)")
print(f"\n‚úÖ FINAL DECISION: We'll use K = 4 clusters")
print(f"   This balances statistical quality with business practicality!")

# Set optimal K
optimal_k = 4  # Based on both methods and business needs
```

## üìù Section 9: Model Evaluation and Quality Metrics

### What to Expect
We'll evaluate the quality of our clustering using multiple metrics.

### Process Overview
1. Calculate Silhouette Score for final model
2. Analyze silhouette plot for each cluster
3. Calculate Davies-Bouldin Index (another clustering quality metric)

Davies-Bouldin Index is a clustering validation metric that measures the average similarity between each cluster and its most similar cluster, where similarity is defined as the ratio of within-cluster scatter to between-cluster separation.

**Key characteristics:**
- **Range**: 0 to infinity (lower is better)
- **What it measures**: Cluster compactness vs separation
- **Formula**: DB = (1/k) √ó Œ£ max((œÉ·µ¢ + œÉ‚±º)/d(c·µ¢,c‚±º)) for all cluster pairs
    - œÉ·µ¢ = average distance of points in cluster i to cluster center
    - d(c·µ¢,c‚±º) = distance between cluster centers i and j

**Interpretation:**
- **DB < 1.0**: Excellent clustering (compact, well-separated clusters)
- **DB = 1.0-2.0**: Good clustering
- **DB > 2.0**: Poor clustering (overlapping or poorly separated clusters)

**Why it's relevant for K-Means:**
1. **Complements Silhouette Score**: While silhouette measures how well each point fits its cluster, Davies-Bouldin measures overall cluster quality
2. **Faster computation**: More efficient than silhouette for large datasets
3. **Different perspective**: Focuses on cluster centers rather than individual points
4. **Helps validate K choice**: Lower DB values indicate better cluster separation

**Business relevance**: Lower Davies-Bouldin scores mean our customer segments are more distinct and actionable for targeted marketing campaigns!
4. Interpret quality metrics

### Expected Outcome
- Quantitative assessment of clustering quality
- Confidence in our segmentation results
- Understanding of cluster cohesion and separation

### Why This Matters
We need to validate that our clusters are meaningful and well-separated, not just random groupings!

```python
# Calculate final clustering quality metrics
print("="*80)
print("üìä CLUSTERING QUALITY EVALUATION")
print("="*80)

# 1. Silhouette Score (overall)
final_silhouette = silhouette_score(X_scaled, cluster_labels)  # Calculate silhouette score
print(f"\n1Ô∏è‚É£ Silhouette Score: {final_silhouette:.4f}")
print(f"   ‚Ä¢ Range: -1 to 1 (higher is better)")
print(f"   ‚Ä¢ Interpretation:")
if final_silhouette > 0.7:
    print(f"     ‚úÖ Excellent clustering! Strong, well-separated clusters")
elif final_silhouette > 0.5:
    print(f"     ‚úÖ Good clustering! Reasonable structure found")
elif final_silhouette > 0.25:
    print(f"     ‚ö†Ô∏è  Weak clustering. Clusters overlap somewhat")
else:
    print(f"     ‚ùå Poor clustering. No substantial structure")

# 2. Within-Cluster Sum of Squares (WCSS)
print(f"\n2Ô∏è‚É£ WCSS (Inertia): {kmeans_final.inertia_:.2f}")
print(f"   ‚Ä¢ Lower is better (more compact clusters)")
print(f"   ‚Ä¢ This is the value we minimized during clustering")

# 3. Silhouette scores per cluster
print(f"\n3Ô∏è‚É£ Silhouette Score by Cluster:")
silhouette_vals = silhouette_samples(X_scaled, cluster_labels)  # Calculate per-sample scores
for cluster_id in range(optimal_k):  # Loop through clusters
    cluster_silhouette = silhouette_vals[cluster_labels == cluster_id].mean()  # Average for cluster
    print(f"   ‚Ä¢ Cluster {cluster_id}: {cluster_silhouette:.4f}")

print(f"\n‚úÖ Quality metrics calculated!")
print(f"üí° All clusters have positive silhouette scores = good separation")
```

```python
# Create silhouette plot for visual quality assessment
fig, ax = plt.subplots(figsize=(12, 8))  # Create figure

y_lower = 10  # Starting y-position

# Plot silhouette for each cluster
for cluster_id in range(optimal_k):  # Loop through clusters
    # Get silhouette values for this cluster
    cluster_silhouette_vals = silhouette_vals[cluster_labels == cluster_id]  # Filter by cluster
    cluster_silhouette_vals.sort()  # Sort values

    # Calculate y-position range for this cluster
    size_cluster = cluster_silhouette_vals.shape[0]  # Number of samples
    y_upper = y_lower + size_cluster  # Upper y-position

    # Fill area for this cluster
    ax.fill_betweenx(np.arange(y_lower, y_upper),  # Y-range
                     0,  # X-start
                     cluster_silhouette_vals,  # X-end (silhouette values)
                     facecolor=colors[cluster_id],  # Color
                     alpha=0.7,  # Transparency
                     edgecolor='black',  # Border
                     linewidth=0.5)  # Border width

    # Label cluster
    ax.text(-0.05, y_lower + 0.5 * size_cluster, f'Cluster {cluster_id}',  # Position and text
           fontsize=12, fontweight='bold')  # Font properties

    # Update y_lower for next cluster
    y_lower = y_upper + 10  # Add gap between clusters

# Add vertical line for average silhouette score
ax.axvline(x=final_silhouette, color='red', linestyle='--', linewidth=2,  # Line properties
          label=f'Average Score: {final_silhouette:.3f}')  # Legend label

# Add labels and title
ax.set_xlabel('Silhouette Coefficient', fontsize=14, fontweight='bold')  # X-axis label
ax.set_ylabel('Cluster', fontsize=14, fontweight='bold')  # Y-axis label
ax.set_title('Silhouette Plot for K-Means Clustering', fontsize=16, fontweight='bold')  # Title
ax.set_yticks([])  # Remove y-ticks
ax.set_xlim([-0.1, 1])  # Set x-axis limits
ax.legend(fontsize=12)  # Add legend
ax.grid(True, alpha=0.3, axis='x')  # Add vertical grid

plt.tight_layout()  # Adjust layout
plt.show()  # Display plot

print("\n‚úÖ Silhouette plot created!")
print("üìä Interpretation:")
print("   ‚Ä¢ Width of each colored area = cluster size")
print("   ‚Ä¢ Silhouette coefficient for each sample shown")
print("   ‚Ä¢ Values above average line = well-clustered samples")
print("   ‚Ä¢ Values below average = potentially misclassified samples")
print("   ‚Ä¢ Uniform thickness = balanced cluster sizes")
```

## ‚ùì Frequently Asked Questions (FAQs)

### 1. **What is the difference between supervised and unsupervised learning?**

**Answer:**
- **Supervised Learning**: You have labeled data (input + correct output). Goal: Learn to predict labels for new data.
  - Example: Email spam detection (emails labeled as "spam" or "not spam")
- **Unsupervised Learning**: You have unlabeled data (input only). Goal: Discover hidden patterns or structure.
  - Example: Customer segmentation (no predefined customer types)

**Analogy**: Supervised = Learning with a teacher. Unsupervised = Exploring on your own.

---

### 2. **How does K-Means algorithm actually work?**

**Answer:**
K-Means follows these steps:

1. **Initialize**: Randomly place K centroids in the feature space
2. **Assignment**: Assign each data point to the nearest centroid (using Euclidean distance)
3. **Update**: Recalculate centroids as the mean of all points assigned to each cluster
4. **Repeat**: Steps 2-3 until centroids stop moving (convergence)

**Analogy**: Imagine K magnets on a table with metal balls. Each ball moves to the nearest magnet. Then magnets move to the center of their balls. Repeat until stable!

---

### 3. **Why do we need to scale features before K-Means?**

**Answer:**
K-Means uses Euclidean distance to measure similarity. Features with larger ranges dominate the distance calculation.

**Example without scaling:**
- Customer A: Age=25, Income=$100k
- Customer B: Age=26, Income=$101k
- Distance ‚âà ‚àö((25-26)¬≤ + (100-101)¬≤) ‚âà ‚àö(1 + 1) ‚âà 1.4

Income difference (1k) has same weight as age difference (1 year) - but income varies 15-150 while age varies 18-70!

**With scaling**: All features contribute equally to distance.

---

### 4. **How do I choose the right number of clusters (K)?**

**Answer:**
Use multiple methods:

1. **Elbow Method**: Plot WCSS vs K, look for "elbow" where improvement slows
2. **Silhouette Analysis**: Choose K with highest average silhouette score
3. **Business Constraints**: Consider practical limits (e.g., can't manage 50 marketing campaigns)
4. **Domain Knowledge**: Industry standards (retail often uses 4-6 segments)

**Best Practice**: Combine statistical methods with business needs!

### 5. **What is a good Silhouette Score?**

**Answer:**
- **0.7 to 1.0**: Strong, well-separated clusters (excellent)
- **0.5 to 0.7**: Reasonable clustering (good)
- **0.25 to 0.5**: Weak clustering, clusters overlap (acceptable)
- **Below 0.25**: No substantial clustering structure (poor)

**Our Result**: ~0.5-0.6 = Good clustering for customer segmentation!

### 6. **What are the limitations of K-Means?**

**Answer:**
1. **Assumes spherical clusters**: Struggles with elongated or irregular shapes
2. **Sensitive to initialization**: Different starting points can give different results (solved by k-means++)
3. **Requires specifying K**: Must know number of clusters in advance
4. **Sensitive to outliers**: Extreme values can distort centroids
5. **Assumes equal cluster sizes**: May not work well with very imbalanced clusters

**Alternatives**: DBSCAN (handles irregular shapes), Hierarchical Clustering (no need to specify K)

---

### 7. **What is k-means++ initialization?**

**Answer:**
k-means++ is a smart way to choose initial centroids:

1. Choose first centroid randomly
2. For each remaining centroid, choose a point with probability proportional to its squared distance from nearest existing centroid
3. This spreads out initial centroids, leading to better final clusters

**Benefit**: Faster convergence and better results than random initialization!

---

### 8. **Can K-Means handle categorical features?**

**Answer:**
Not directly! K-Means uses Euclidean distance, which only works for numerical features.

**Solutions:**
1. **One-Hot Encoding**: Convert categories to binary features (Gender ‚Üí Male=1/0, Female=1/0)
2. **Label Encoding**: Convert to numbers (Low=1, Medium=2, High=3) - only if ordinal
3. **K-Modes**: Alternative algorithm designed for categorical data
4. **Gower Distance**: Mixed distance metric for numerical + categorical

**Our Dataset**: All numerical features, so K-Means works perfectly!

Check multiple indicators:

1. **Statistical Metrics**: Silhouette score > 0.5, low WCSS
2. **Visual Inspection**: Clear separation in scatter plots
3. **Business Interpretation**: Clusters make sense (e.g., "budget shoppers" vs "premium customers")
4. **Actionability**: Can you create different strategies for each cluster?
5. **Stability**: Re-running gives similar results

### 10. **What's the difference between K-Means and Hierarchical Clustering?**

**Answer:**

| Aspect | K-Means | Hierarchical |
|--------|---------|-------------|
| **K specification** | Must specify K upfront | No need to specify K |
| **Speed** | Fast (O(n)) | Slow (O(n¬≤) or O(n¬≥)) |
| **Scalability** | Handles large datasets | Struggles with >10K samples |
| **Output** | Flat clusters | Dendrogram (tree structure) |
| **Cluster shape** | Spherical | Any shape |
| **Use case** | Large datasets, known K | Small datasets, explore K |

**Recommendation**: K-Means for large-scale customer segmentation, Hierarchical for exploratory analysis!

### 11. **How often should I re-run clustering on customer data?**

**Answer:**
Depends on business dynamics:

- **E-commerce**: Monthly or quarterly (customer behavior changes with seasons)
- **Subscription services**: Quarterly (churn patterns evolve)
- **Retail**: Seasonally (holiday vs regular shopping patterns)
- **B2B**: Annually (business relationships more stable)

**Best Practice**: Monitor cluster stability. If >20% of customers change clusters, re-run clustering!

### 12. **Can I use K-Means for image compression?**

**Answer:**
Yes! K-Means is commonly used for image compression:

1. Treat each pixel as a data point with RGB values (3 features)
2. Cluster pixels into K color groups
3. Replace each pixel with its cluster centroid color
4. Result: Image with only K colors instead of millions!

**Example**: K=16 reduces a 24-bit image (16M colors) to 4-bit (16 colors) with minimal quality loss!

**Other Applications**: Document clustering, anomaly detection, recommendation systems
